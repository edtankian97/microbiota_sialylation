```r
#get unique species

all_gtdbtk_bac120_summary_colon03 <- read_delim("all_colon03_summ_gtdb_arch_bac.tsv", 
                                                delim = "\t", escape_double = FALSE, 
                                                trim_ws = TRUE)
head(all_gtdbtk_bac120_summary_colon03)
```

```r
# Upload result of checkm filtering

checkm_summary_colon03 <- read_delim("quality_report_bin_filtered_ID.tsv",
                                     delim = "\t", escape_double = FALSE, 
                                     trim_ws = TRUE, col_names = "user_genome")

head(checkm_summary_colon03)
# Filter ID based on checkm result
gtdbtk_bac120_summary_colon03 <- semi_join(all_gtdbtk_bac120_summary_colon03, checkm_summary_colon03, by= "user_genome")
```

```r
#get species name
gtdbtk_classify_colon03 <- gtdbtk_bac120_summary_colon03[,"classification"]

gtdbtk_species_colon03 <- gtdbtk_classify_colon03 %>%
  mutate(species = str_extract(classification, "(?<=s__).*"))

gtdbtk_species_colon03$species <- trimws(gtdbtk_species_colon03$species)
gtdbtk_species_colon03$species[gtdbtk_species_colon03$species == ""] <- NA

unique(gtdbtk_species_colon03$species)

#count species match
table(gtdbtk_species_colon03$species)
```

```r
count_gtdb_colon03 <- gtdbtk_species_colon03 %>%
  count(species, name = "occurrences")

head(count_gtdb_colon03)


unique_gtdbtk_species_colon03 <- gtdbtk_species_colon03 %>% distinct(species, .keep_all = TRUE)

unique_gtdbtk_species_colon03_ID <- unique_gtdbtk_species_colon03[,"species"]
```

```r
#Save output
write_tsv(unique_gtdbtk_species_colon03_ID, "GTDB_results/unique_gtdbtk_species_colon03_ID.tsv")
```

```r
################################################################################################################
#Division of disease versus Healthy


# Remove everything after the first "-"
gtdbtk_bac120_summary_colon03$ID <- sub("_.*", "", gtdbtk_bac120_summary_colon03$user_genome)

#get ID
gtdbtk_colon03_ID_name <- gtdbtk_bac120_summary_colon03[,"ID"]

```

```r
#save file
write_tsv(gtdbtk_colon03_ID_name, "GTDB_results/gtdbtk_colon03_ERR_ID_name.tsv")
```

```r
#Upload data with health and disease ID

#case
Study_3_cancer_case_Run_ID <- read_csv("all_cancer_study3_ID")

head(Study_3_cancer_case_Run_ID)
```

```r
#control

Study_3_cancer_control_Run_ID <- read_csv("all_control_study3_ID.tsv")
head(Study_3_cancer_control_Run_ID)
```

```r
#match and see how many match
colnames(gtdbtk_colon03_ID_name) <- "Run"

#for case
match_colon03_case <- semi_join(gtdbtk_colon03_ID_name,Study_3_cancer_case_Run_ID, by= "Run")
nrow(match_colon03_case)

#get unique IDs
match_colon03_case_unique <- match_colon03_case %>% distinct(Run)
nrow(match_colon03_case_unique)
```

```r
#for control
match_colon03_control <- semi_join(gtdbtk_colon03_ID_name,Study_3_cancer_control_Run_ID, by= "Run")
nrow(match_colon03_control)
#get unique IDs

match_colon03_control_unique <- match_colon03_control %>% distinct(Run)
nrow(match_colon03_control_unique)
```

```r
#Final table 

cat_gtdk_ID <- merge(gtdbtk_bac120_summary_colon03,gtdbtk_species_colon03, by="classification")

cat_gtdk_ID_col <- cat_gtdk_ID[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species")]
head(cat_gtdk_ID)
nrow(cat_gtdk_ID)
```

```r
cat_gtdk_ID_unique <- cat_gtdk_ID%>% distinct(user_genome,.keep_all=TRUE)
nrow(cat_gtdk_ID_unique)
head(cat_gtdk_ID_unique)
```

```r
cat_gtdk_ID_col <- cat_gtdk_ID_unique[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species", "user_genome")]


cat_colon03_preview <- merge(cat_gtdk_ID_col,count_gtdb_colon03, by= "species")
head(cat_colon03_preview)
```

```r
# handling with control and disease column

match_colon03_control$State <- "Control"
match_colon03_case$State <- "Disease"

# Change column name

colnames(match_colon03_control)[colnames(match_colon03_control) == "Run"] <- "ID"
colnames(match_colon03_case)[colnames(match_colon03_case) == "Run"] <- "ID" 
```

```r
all_state <- rbind(match_colon03_control,match_colon03_case)

cat_colon03_final <- merge(cat_colon03_preview,all_state, by= "ID")
head(cat_colon03_final)
nrow(cat_colon03_final)
```

```r
cat_colon03_final_end <- cat_colon03_final %>% distinct(user_genome, .keep_all = TRUE)
nrow(cat_colon03_final_end)
```

```r
write_tsv(cat_colon03_final_end, "GTDB_results/final_table_gtdb_for_paper_colon03.tsv")
write_tsv(count_gtdb_colon03, "GTDB_results/species_occurrence.tsv")
```

```r
# Classify species
classification <- cat_colon03_final_end %>%
  filter(!is.na(species), !is.na(State)) %>%
  group_by(species) %>%
  summarise(
    Group_classification = case_when(
      n_distinct(State) == 1 & first(State) == "Disease" ~ "Disease",
      n_distinct(State) == 1 & first(State) == "Control" ~ "Control",
      n_distinct(State) > 1 ~ "Control/Disease"
    ),
    .groups = "drop"
  )

# Join  and apply rule for Species == NA
cat_colon03_final_end_df <- cat_colon03_final_end %>%
  left_join(classification, by = "species") %>%
  mutate(
    Group_classification = ifelse(
      is.na(species),
      State,
      Group_classification
    )
  )

head(cat_colon03_final_end_df)
```

```r
write_tsv(cat_colon03_final_end_df,"GTDB_results/gtdb_final_table_with_condition_colon03_corrected.tsv")
```

```r
cat_colon03_final_end_df_col <- cat_colon03_final_end_df[ ,c("species", "occurrences", "Group_classification")]

cat_colon03_final_end_df_col_unique <- cat_colon03_final_end_df_col %>% distinct(species, .keep_all = TRUE)

head(cat_colon03_final_end_df_col_unique)
nrow(cat_colon03_final_end_df_col_unique)
```
