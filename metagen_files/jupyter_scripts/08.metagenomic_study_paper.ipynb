```r
#colon cancer study

#case 01 - France
Study_1_metadata_France_colon_rectar <- read_csv("Metadata/Study_1_metadata_France_colon_rectar.csv")
head(Study_1_metadata_France_colon_rectar)
```

```r
statics <- as.data.frame(table(Study_1_metadata_France_colon_rectar$host_subject_id))

colnames(statics)[colnames(statics) == "Var1" ] <- "host_subject_id"
connected_meta <- inner_join(statics, Study_1_metadata_France_colon_rectar, by = "host_subject_id")

#select specific collumns
final_meta_study1 <- connected_meta[, c("host_subject_id", "Freq", "diagnosis")]

#get uniques lines
final_meta_study_1_unique <- distinct(final_meta_study1, .keep_all = TRUE)

#filter by country

study1_france <- final_meta_study_1_unique %>%
  filter(str_detect(host_subject_id, "FR"))

#filter France by condition
study1_france_cancer <- study1_france[study1_france$diagnosis == "Cancer", ]
study1_france_small_adeno <-  study1_france[study1_france$diagnosis == "Small adenoma", ]
study1_france_large_adeno <-  study1_france[study1_france$diagnosis == "Large adenoma", ]
study1_france_normal <-  study1_france[study1_france$diagnosis == "Normal", ]

#take ID from cancer - France
france_cancer_ID <- inner_join(study1_france_cancer, Study_1_metadata_France_colon_rectar, by = "host_subject_id")
#take collumns
france_cancer_get_ID <- france_cancer_ID %>% select(Run)

write_tsv(france_cancer_get_ID, "france_cancer_get_ID.tsv")

#take ID from small adeno - France
france_small_ID <- inner_join(study1_france_small_adeno, Study_1_metadata_France_colon_rectar, by = "host_subject_id")
#take collumns
france_small_get_ID <- france_small_ID %>% select(Run)

write_tsv(france_small_get_ID, "france_small_adenoma_get_ID.tsv")

#take ID from large adeno - France
france_large_ID <- inner_join(study1_france_large_adeno, Study_1_metadata_France_colon_rectar, by = "host_subject_id")
#take collumns
france_large_get_ID <- france_large_ID%>% select(Run)

write_tsv(france_large_get_ID, "france_large_adenoma_get_ID.tsv")

#take ID from normal - France
france_normal_ID <- inner_join(study1_france_normal, Study_1_metadata_France_colon_rectar, by = "host_subject_id")
#take collumns
france_normal_get_ID <- france_normal_ID%>% select(Run)

write_tsv(france_normal_get_ID, "france_normal_get_ID.tsv")
```

```r
############################ Study3 - colon cancer


Study_3_colon_cancer_metadata <- read_csv("Metadata/Study_3_colon_cancer_metadata.csv")
head(Study_3_colon_cancer_metadata)
```

```r
#import supplementary data to match sample IDs
library(readxl)

study3_cancer_suplem_data <-read_xlsx("Metadata/Study3_colon_cancer_supplementary_data.xlsx")
head(study3_cancer_suplem_data)
```

```r
unique(study3_cancer_suplem_data$State[duplicated(study3_cancer_suplem_data$State)])
#join data

#select specific collumns
study3_cancer_suplem_data_columm <- study3_cancer_suplem_data[, c("Sample ID", "State","Gender (1:male, 2:female)")]

colnames(study3_cancer_suplem_data_columm)[1] <- "sample_name"
#separate controls by male and female
study3_cancer_suplem_data_control_male <- study3_cancer_suplem_data_columm%>%filter(State=="controls" & 
                                                                                      `Gender (1:male, 2:female)` =="1")

study3_cancer_suplem_data_control_female <- study3_cancer_suplem_data_columm%>%filter(State=="controls" & 
                                                                                        `Gender (1:male, 2:female)` =="2")


#separate advanced adenoma by male and female

study3_cancer_suplem_data_adenoma_male <- study3_cancer_suplem_data_columm%>%filter(State=="advanced adenoma" & 
                                                                                        `Gender (1:male, 2:female)` =="1")


study3_cancer_suplem_data_adenoma_female <- study3_cancer_suplem_data_columm%>%filter(State=="advanced adenoma" & 
                                                                                      `Gender (1:male, 2:female)` =="2")

#separate carcinoma by male and female

study3_cancer_suplem_data_carcinoma_male <- study3_cancer_suplem_data_columm%>%filter(State=="carcinoma" & 
                                                                                        `Gender (1:male, 2:female)` =="1")

study3_cancer_suplem_data_carcinoma_female <- study3_cancer_suplem_data_columm%>%filter(State=="carcinoma" & 
                                                                                        `Gender (1:male, 2:female)` =="2")


#match files in control
control_male_study3 <- inner_join(study3_cancer_suplem_data_control_male,
                                  Study_3_colon_cancer_metadata, by = "sample_name")

control_male_study3_run <- control_male_study3[,4]

write_tsv(control_male_study3_run, "control_male_study3.tsv")

#female

control_female_study3 <- inner_join(study3_cancer_suplem_data_control_female,
                                  Study_3_colon_cancer_metadata, by = "sample_name")

control_female_study3_run <- control_female_study3[,4]
write_tsv(control_female_study3_run,"control_female_study3.tsv")

#match files in advanced adenoma

#male
adenoma_male_study3 <- inner_join(study3_cancer_suplem_data_adenoma_male,
                                  Study_3_colon_cancer_metadata, by = "sample_name")

#male Run columm
adenoma_male_study3_run <- adenoma_male_study3[,4]
write_tsv(adenoma_male_study3_run, "adenoma_male_study3.tsv")

#female
adenoma_female_study3 <- inner_join(study3_cancer_suplem_data_adenoma_female,
                                  Study_3_colon_cancer_metadata, by = "sample_name")

adenoma_female_study3_run <- adenoma_female_study3[,4]
write_tsv(adenoma_female_study3_run, "adenoma_female_study3.tsv")


#match files in carcinoma
#male
carcinoma_male_study3 <- inner_join(study3_cancer_suplem_data_carcinoma_male,
                                  Study_3_colon_cancer_metadata, by = "sample_name")

carcinoma_male_study3_run <- carcinoma_male_study3[,4]
write_tsv(carcinoma_male_study3_run, "carcinoma_male_study3.tsv")



#female
carcinoma_female_study3 <- inner_join(study3_cancer_suplem_data_carcinoma_female,
                                    Study_3_colon_cancer_metadata, by = "sample_name")

carcinoma_female_study3_run <- carcinoma_female_study3[,4]
write_tsv(carcinoma_female_study3_run, "carcinoma_female_study3.tsv")
```

```r
#################################################################################################
# Aterosclerosis - Study 01 - Chinese cohort


library(readr)
atero_study01 <- read_csv("Metadata/atero_study01.csv")
head(atero_study01)
```

```r
atero_supple_data <- read_xlsx("Metadata/atero_study01.xlsx", skip = 2, col_names = TRUE)
head(atero_supple_data)
```

```r
#control

#male
atero_male_control_suplem <- atero_supple_data%>%filter(str_detect(`Sample ID`, "N") & 
                                                          Gender =="male")

#separate column and match with main metadata to get Run ID

atero_male_control_suplem_ID <- atero_male_control_suplem[,1]
colnames(atero_male_control_suplem_ID) <- "Sample_name"

atero_male_final_ID <- inner_join(atero_male_control_suplem_ID,atero_study01, by= "Sample_name")

#separate desided column with ID run
atero_male_control_ID <- atero_male_final_ID[,2]
write_tsv(atero_male_control_ID, "atero_study1_male_control_ID")

#female
atero_female_control_suplem <- atero_supple_data%>%filter(str_detect(`Sample ID`, "N") & 
                                                          Gender =="female")

#separate column and match with main metadata to get Run ID
atero_female_control_suplem_ID <- atero_female_control_suplem[,1]
colnames(atero_female_control_suplem_ID) <- "Sample_name"

atero_female_final_ID <- inner_join(atero_female_control_suplem_ID,atero_study01, by= "Sample_name")

#separate desided column with ID run
atero_female_control_ID <- atero_female_final_ID[,2]
write_tsv(atero_female_control_ID, "atero_study1_female_control_ID")


#disease ID

#male
atero_male_disease_suplem <- atero_supple_data%>%filter(str_detect(`Sample ID`, "ZSL") & 
                                                          Gender =="male")

#separate column and match with main metadata to get Run ID

atero_male_disease_suplem_ID <- atero_male_disease_suplem[,1]
colnames(atero_male_disease_suplem_ID) <- "Sample_name"

atero_disease_male_final_ID <- inner_join(atero_male_disease_suplem_ID,atero_study01, by= "Sample_name")

#separate desided column with ID run
atero_male_disease_ID <- atero_disease_male_final_ID[,2]
write_tsv(atero_male_disease_ID, "atero_study1_male_disease_ID")


#female
atero_female_disease_suplem <- atero_supple_data%>%filter(str_detect(`Sample ID`, "ZSL") & 
                                                          Gender =="female")

#separate column and match with main metadata to get Run ID

atero_female_disease_suplem_ID <- atero_female_disease_suplem[,1]
colnames(atero_female_disease_suplem_ID) <- "Sample_name"

atero_disease_female_final_ID <- inner_join(atero_female_disease_suplem_ID,atero_study01, by= "Sample_name")

#separate desided column with ID run
atero_female_disease_ID <- atero_disease_female_final_ID[,2]
write_tsv(atero_female_disease_ID, "atero_study1_female_disease_ID")
```
