```r
##################################################################################################
#join all data


#extract GCF ID
CMP_interpro_distinct_atero01$GCF_ID <- str_extract(CMP_interpro_distinct_atero01$Sequence,
                                                    ".*?\\.faa")

#get CMP GCF ID
get_CMP_final_interpro_ID_atero01 <- CMP_interpro_distinct_atero01[,"GCF_ID"]


#pm1088
pm1088_interpro_distinct_atero01$GCF_ID <- str_extract(pm1088_interpro_distinct_atero01$Sequence,
                                                       ".*?\\.faa")
get_pm1088_final_interpro_ID_atero01 <- pm1088_interpro_distinct_atero01[,"GCF_ID"]


#NeuS
neuS_interpro_distinct_atero01$GCF_ID <- str_extract(neuS_interpro_distinct_atero01$Sequence,
                                                     ".*?\\.faa")
get_neuS_final_interpro_ID_atero01 <- neuS_interpro_distinct_atero01[,"GCF_ID"]

#NeuS PFAM
neuS_PFAM_interpro_distinct_atero01$GCF_ID <- str_extract(neuS_PFAM_interpro_distinct_atero01$Sequence,
                                                          ".*?\\.faa")
get_neuS_PFAM_final_interpro_ID_atero01 <- neuS_PFAM_interpro_distinct_atero01[,"GCF_ID"]

#Lst 
Lst_interpro_distinct_atero01$GCF_ID <- str_extract(Lst_interpro_distinct_atero01$Sequence,
                                                    ".*?\\.faa")
get_Lst_final_interpro_ID_atero01 <- Lst_interpro_distinct_atero01[,"GCF_ID"]

#pm1088 - PFAM
pm1088_PFAM_interpro_distinct_atero01$GCF_ID <- str_extract(pm1088_PFAM_interpro_distinct_atero01$Sequence,
                                                            ".*?\\.faa")
get_pm1088_PFAM_final_interpro_ID_atero01 <- pm1088_PFAM_interpro_distinct_atero01[,"GCF_ID"]
```

```r
#join all sialyltransferase
all_sialyltrans_ID_atero01 <- rbind(get_neuS_final_interpro_ID_atero01,get_Lst_final_interpro_ID_atero01,
                                    get_neuS_PFAM_final_interpro_ID_atero01,get_pm1088_PFAM_final_interpro_ID_atero01,
                                    get_pm1088_final_interpro_ID_atero01)

all_sialyltrans_ID_unique_atero01 <-all_sialyltrans_ID_atero01 %>% distinct(GCF_ID)
```

```r
#After filtration, see what are complete pathway that match with CMP
complete_sialylation_interpro_atero01 <- semi_join(get_CMP_final_interpro_ID_atero01, all_sialyltrans_ID_unique_atero01, by="GCF_ID")

write_tsv(complete_sialylation_interpro_atero01, 
          "Interpro_results/complete_sialylation_interpro_filtration_final_atero01_metagen_update.tsv")

#get bins
get_bins_atero01 <- semi_join(CMP_interpro_distinct_atero01, complete_sialylation_interpro_atero01, by="GCF_ID")
get_bins_atero01_ID <- get_bins_atero01[,"Sequence"]


get_bins_atero01_ID <- get_bins_atero01_ID %>%
  mutate(Sequence = Sequence %>%
           str_remove("^PROKKA_") %>%
           str_replace("\\.faa.*$", ".fa"))
```
