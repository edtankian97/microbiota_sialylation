```r
# Upload libraries
library(readr)
library(dplyr)
library(stringr)
```

```r
setwd("/mnt/l/atero1_hmmer_backup")
```

```r
#get unique species

all_gtdbtk_bac120_summary_atero01 <- read_delim("all_atero01_summ_gtdb_arch_bac.tsv", 
                                                delim = "\t", escape_double = FALSE, 
                                                trim_ws = TRUE)
head(all_gtdbtk_bac120_summary_atero01)
```

```r
# Upload result of checkm filtering

checkm_summary_atero01 <- read_delim("quality_report_bin_filtered_ID.tsv",
                                     delim = "\t", escape_double = FALSE, 
                                     trim_ws = TRUE, col_names = "user_genome")

head(checkm_summary_atero01)
```

```r
# Filter ID based on checkm result
gtdbtk_bac120_summary_atero01 <- semi_join(all_gtdbtk_bac120_summary_atero01, checkm_summary_atero01, by= "user_genome")


#get species name
gtdbtk_classify_atero01 <- gtdbtk_bac120_summary_atero01[,"classification"]

gtdbtk_species_atero01 <- gtdbtk_classify_atero01 %>%
  mutate(species = str_extract(classification, "(?<=s__).*"))

gtdbtk_species_atero01$species <- trimws(gtdbtk_species_atero01$species)
gtdbtk_species_atero01$species[gtdbtk_species_atero01$species == ""] <- NA

unique(gtdbtk_species_atero01$species)
```

```r
#count species match
table(gtdbtk_species_atero01$species)

count_gtdb_atero01 <- gtdbtk_species_atero01 %>%
  count(species, name = "occurrences")

head(count_gtdb_atero01)
```

```r
unique_gtdbtk_species_atero01 <- gtdbtk_species_atero01 %>% distinct(species, .keep_all = TRUE)

unique_gtdbtk_species_atero01_ID <- unique_gtdbtk_species_atero01[,"species"]
```

```r
#Save output
write_tsv(unique_gtdbtk_species_atero01_ID, "GTDB_results/unique_gtdbtk_species_atero01_ID.tsv")
```

```r
################################################################################################################
#Division of disease versus Healthy


# Remove everything after the first "-"
gtdbtk_bac120_summary_atero01$ID <- sub("_.*", "", gtdbtk_bac120_summary_atero01$user_genome)

#get ID
gtdbtk_atero01_ID_name <- gtdbtk_bac120_summary_atero01[,"ID"]
```

```r
#save file
write_tsv(gtdbtk_atero01_ID_name, "GTDB_results/gtdbtk_atero01_ERR_ID_name.tsv")
```

```r
#Upload data with health and disease ID

#case
Study_1_atero_case_Run_ID <- read_csv("atero_study1_disease_ID")

head(Study_1_atero_case_Run_ID)
```

```r
#control

Study_1_atero_control_Run_ID <- read_csv("atero_study1_control_ID")
head(Study_1_atero_control_Run_ID)


#match and see how many match
colnames(gtdbtk_atero01_ID_name) <- "Run"
```

```r
#for case
match_atero01_case <- semi_join(gtdbtk_atero01_ID_name,Study_1_atero_case_Run_ID, by= "Run")
nrow(match_atero01_case)

#get unique IDs
match_atero01_case_unique <- match_atero01_case %>% distinct(Run)
nrow(match_atero01_case_unique)
```

```r
#for control
match_atero01_control <- semi_join(gtdbtk_atero01_ID_name,Study_1_atero_control_Run_ID, by= "Run")
nrow(match_atero01_control)
#get unique IDs

match_atero01_control_unique <- match_atero01_control %>% distinct(Run)
nrow(match_atero01_control_unique)
```

```r
#Final table 

cat_gtdk_ID <- merge(gtdbtk_bac120_summary_atero01,gtdbtk_species_atero01, by="classification")

cat_gtdk_ID_col <- cat_gtdk_ID[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species")]
head(cat_gtdk_ID)
nrow(cat_gtdk_ID)
```

```r
cat_gtdk_ID_unique <- cat_gtdk_ID%>% distinct(user_genome,.keep_all=TRUE)
nrow(cat_gtdk_ID_unique)
head(cat_gtdk_ID_unique)
```

```r
cat_gtdk_ID_col <- cat_gtdk_ID_unique[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species", "user_genome")]


cat_atero01_preview <- merge(cat_gtdk_ID_col,count_gtdb_atero01, by= "species")
head(cat_atero01_preview)
```

```r
# handling with control and disease column

match_atero01_control$State <- "Control"
match_atero01_case$State <- "Disease"

# Change column name

colnames(match_atero01_control)[colnames(match_atero01_control) == "Run"] <- "ID"
colnames(match_atero01_case)[colnames(match_atero01_case) == "Run"] <- "ID" 

all_state <- rbind(match_atero01_control,match_atero01_case)

cat_atero01_final <- merge(cat_atero01_preview,all_state, by= "ID")
head(cat_atero01_final)
nrow(cat_atero01_final)
```

```r
cat_atero01_final_end <- cat_atero01_final %>% distinct(user_genome, .keep_all = TRUE)
head(cat_atero01_final_end)
nrow(cat_atero01_final_end)
```

```r
write_tsv(cat_atero01_final_end, "GTDB_results/final_table_gtdb_for_paper_atero01.tsv")
write_tsv(count_gtdb_atero01, "GTDB_results/species_occurrence.tsv")
```

```r
# Classify species
classification <- cat_atero01_final_end %>%
  filter(!is.na(species), !is.na(State)) %>%
  group_by(species) %>%
  summarise(
    Group_classification = case_when(
      n_distinct(State) == 1 & first(State) == "Disease" ~ "Disease",
      n_distinct(State) == 1 & first(State) == "Control" ~ "Control",
      n_distinct(State) > 1 ~ "Control/Disease"
    ),
    .groups = "drop"
  )

# Join  and apply rule for Species == NA
cat_atero01_final_end_df <- cat_atero01_final_end %>%
  left_join(classification, by = "species") %>%
  mutate(
    Group_classification = ifelse(
      is.na(species),
      State,
      Group_classification
    )
  )

head(cat_atero01_final_end_df)
```

```r
write_tsv(cat_atero01_final_end_df,"GTDB_results/gtdb_final_table_with_condition_atero01_corrected.tsv")
```

```r
cat_atero01_final_end_df_col <- cat_atero01_final_end_df[ ,c("species", "occurrences", "Group_classification")]

cat_atero01_final_end_df_col_unique <- cat_atero01_final_end_df_col %>% distinct(species, .keep_all = TRUE)
```

```r
# Output table for final paper
head(cat_atero01_final_end_df_col_unique)
nrow(cat_atero01_final_end_df_col_unique)
```

```r
write_tsv(cat_atero01_final_end_df_col_unique, "GTDB_results/count_ocurrence_w_condition_for_paper_atero01.tsv")
```
