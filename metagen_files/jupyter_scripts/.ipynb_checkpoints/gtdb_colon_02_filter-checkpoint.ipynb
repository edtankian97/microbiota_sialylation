```r
# Upload result of checkm filtering

checkm_summary_colon02 <- read_delim("quality_report_bin_filtered_ID.tsv",
                                     delim = "\t", escape_double = FALSE, 
                                     trim_ws = TRUE, col_names = "user_genome")
head(checkm_summary_colon02)
```

```r
# Filter ID based on checkm result
gtdbtk_bac120_summary_colon02 <- semi_join(all_gtdbtk_bac120_summary_colon02, checkm_summary_colon02, by= "user_genome")


#get species name
gtdbtk_classify_colon02 <- gtdbtk_bac120_summary_colon02[,"classification"]

gtdbtk_species_colon02 <- gtdbtk_classify_colon02 %>%
  mutate(species = str_extract(classification, "(?<=s__).*"))

gtdbtk_species_colon02$species <- trimws(gtdbtk_species_colon02$species)
gtdbtk_species_colon02$species[gtdbtk_species_colon02$species == ""] <- NA

unique(gtdbtk_species_colon02$species)
```

```r
#count species match
table(gtdbtk_species_colon02$species)

count_gtdb_colon02 <- gtdbtk_species_colon02 %>%
  count(species, name = "occurrences")

View(count_gtdb_colon02)


unique_gtdbtk_species_colon02 <- gtdbtk_species_colon02 %>% distinct(species, .keep_all = TRUE)

unique_gtdbtk_species_colon02_ID <- unique_gtdbtk_species_colon02[,"species"]
```

```r
write_tsv(unique_gtdbtk_species_colon02_ID, "GTDB_results/unique_gtdbtk_species_colon02_ID.tsv")
```

```r
################################################################################################################
#Division of disease versus Healthy


# Remove everything after the first "-"
gtdbtk_bac120_summary_colon02$ID <- sub("_.*", "", gtdbtk_bac120_summary_colon02$user_genome)

#get ID
gtdbtk_colon02_ID_name <- gtdbtk_bac120_summary_colon02[,"ID"]
head(gtdbtk_colon02_ID_name)
```

```r
#save file
write_tsv(gtdbtk_colon02_ID_name, "GTDB_results/gtdbtk_colon02_ID_name.tsv")
```

```r
#Upload data with health and disease ID. Put PATH to Metadata folder

#case
Study_2_china_cancer_case_Run_ID <- read_csv("Metadata/Study_2_china_cancer_case_Run_ID.tsv")

head(Study_2_china_cancer_case_Run_ID)
```

```r
#control

Study_2_china_cancer_control_Run_ID <- read_csv("Metadata/Study_2_china_cancer_control_Run_ID.tsv")
head(Study_2_china_cancer_control_Run_ID)


#match and see how many match
colnames(gtdbtk_colon02_ID_name) <- "Run"
```

```r
#for case
match_colon02_case <- semi_join(gtdbtk_colon02_ID_name,Study_2_china_cancer_case_Run_ID, by= "Run")
nrow(match_colon02_case)

#get unique IDs
match_colon02_case_unique <- match_colon02_case %>% distinct(Run)
nrow(match_colon02_case_unique)
```

```r
#for control
match_colon02_control <- semi_join(gtdbtk_colon02_ID_name,Study_2_china_cancer_control_Run_ID, by= "Run")
nrow(match_colon02_control)

#get unique IDs
match_colon02_control_unique <- match_colon02_control %>% distinct(Run)
nrow(match_colon02_control_unique)
```

```r
#Final table 

cat_gtdk_ID <- merge(gtdbtk_bac120_summary_colon02,gtdbtk_species_colon02, by="classification")

cat_gtdk_ID_col <- cat_gtdk_ID[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species")]
nrow(cat_gtdk_ID)

cat_gtdk_ID_unique <- cat_gtdk_ID%>% distinct(user_genome,.keep_all=TRUE)
nrow(cat_gtdk_ID_unique)
head(cat_gtdk_ID_unique)
```

```r
cat_gtdk_ID_col <- cat_gtdk_ID_unique[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species", "user_genome")]


cat_colon02_preview <- merge(cat_gtdk_ID_col,count_gtdb_colon02, by= "species")
head(cat_colon02_preview)
```

```r
# handling with control and disease column

match_colon02_control$State <- "Control"
match_colon02_case$State <- "Disease"

# Change column name

colnames(match_colon02_control)[colnames(match_colon02_control) == "Run"] <- "ID"
colnames(match_colon02_case)[colnames(match_colon02_case) == "Run"] <- "ID" 

all_state <- rbind(match_colon02_control,match_colon02_case)

cat_colon02_final <- merge(cat_colon02_preview,all_state, by= "ID")
head(cat_colon02_final)
```

```r
cat_colon02_final_end <- cat_colon02_final %>% distinct(user_genome, .keep_all = TRUE)
nrow(cat_colon02_final_end)
```

```r
#save output. Put PATH to Study02_colon_cancer folder
write_tsv(cat_colon02_final_end, "GTDB_results/final_table_gtdb_for_paper_colon02.tsv")
write_tsv(count_gtdb_colon02, "GTDB_results/species_occurrence.tsv")
```

```r
# Classify species
classification <- cat_colon02_final_end %>%
  filter(!is.na(species), !is.na(State)) %>%
  group_by(species) %>%
  summarise(
    Group_classification = case_when(
      n_distinct(State) == 1 & first(State) == "Disease" ~ "Disease",
      n_distinct(State) == 1 & first(State) == "Control" ~ "Control",
      n_distinct(State) > 1 ~ "Control/Disease"
    ),
    .groups = "drop"
  )

# Join  and apply rule for Species == NA
cat_colon02_final_end_df <- cat_colon02_final_end %>%
  left_join(classification, by = "species") %>%
  mutate(
    Group_classification = ifelse(
      is.na(species),
      State,
      Group_classification
    )
  )

head(cat_colon02_final_end_df)
```

```r
write_tsv(cat_colon02_final_end_df,"GTDB_results/gtdb_final_table_with_condition_colon02_corrigido.tsv")
```

```r
cat_colon02_final_end_df_col <- cat_colon02_final_end_df[ ,c("species", "occurrences", "Group_classification")]

cat_colon02_final_end_df_col_unique <- cat_colon02_final_end_df_col %>% distinct(species, .keep_all = TRUE)
```

```r
write_tsv(cat_colon02_final_end_df_col_unique, "GTDB_results/count_ocurrence_w_condition_for_paper_colon02.tsv")
```
