```r
matched_CMP_ID_interpro_colon02 <- read_delim("Interpro_results/colon02_CMP_ID_sequence_for_Interpro.tsv_retrieved_now_results.tsv", 
                                         delim = "\t", escape_double = FALSE, 
                                         col_names = FALSE, trim_ws = TRUE)


column_names_interpro <-c("Sequence", "md5", "unkown", "Bank", "signature", "annotation", "align_begin",
                          "align_end", "e-value","Hit", "date", "Interpro_code", "annotation_two", "unkown2", "unkown3")

setnames(matched_CMP_ID_interpro_colon02, column_names_interpro)

unique(matched_CMP_ID_interpro_colon02$annotation)

patterns_of_teste_CMP <- c("TIGR00466","PTHR42866","NF003952", "cd02517", "MF_00057", "TIGR03584") #remove only other nonulosonic citidyltransferase
```

```r
# 1. Find all Sequence IDs containing any of the patterns
bad_ids <- matched_CMP_ID_interpro_colon02 %>%
  filter(grepl(paste(patterns_of_teste_CMP, collapse = "|"), signature)) %>%
  pull(Sequence) %>%
  unique()

# 2. Remove those IDs entirely
filtered_CMP_interpro_rm_colon02 <- matched_CMP_ID_interpro_colon02 %>%
  filter(!(Sequence %in% bad_ids))

patterns <- c("PF02348","SSF53448","G3DSA:3.90.550.10","PTHR21485","cd02513")

filtered_CMP_interpro_colon02 <- filtered_CMP_interpro_rm_colon02 %>%
  filter(signature %in% patterns)

head(filtered_CMP_interpro_colon02)

CMP_interpro_distinct_colon02 <- filtered_CMP_interpro_colon02%>% distinct(Sequence, .keep_all = TRUE)
```

```r
#pm1088

pm1088_Interpro_data_colon02 <- read_delim("Interpro_results/colon02_pm0188_ID_sequence_for_Interpro.tsv_retrieved_now_results.tsv", 
                                      delim = "\t", escape_double = FALSE, 
                                      col_names = FALSE, trim_ws = TRUE)
head(pm1088_Interpro_data_colon02)

setnames(pm1088_Interpro_data_colon02, column_names_interpro)


pm1088_interpro_annot_unique_colon02 <- unique(pm1088_Interpro_data_colon02$annotation)
pm1088_interpro_annot_unique_colon02

patterns_off_pm1088 <- c("PTHR30160","PF01075","G3DSA:2.60.40.10","cd14948","G3DSA:3.40.50.2000", "PF22352") #remove non pm0188 sequences
```

```r
# 1. Find all Sequence IDs containing any of the patterns
bad_ids_pm1088_colon02 <- pm1088_Interpro_data_colon02 %>%
  filter(grepl(paste(patterns_off_pm1088, collapse = "|"), signature)) %>%
  pull(Sequence) %>%
  unique()

# 2. Remove those IDs entirely
filtered_pm1088_interpro_colon02 <- pm1088_Interpro_data_colon02 %>%
  filter(!(Sequence %in% bad_ids_pm1088_colon02))


patterns_pm1088 <- c("PF11477")

filtered_pm1088_interpro_colon02_test <- filtered_pm1088_interpro_colon02 %>%
  filter(signature %in% patterns_pm1088)

pm1088_interpro_distinct_colon02 <- filtered_pm1088_interpro_colon02_test  %>% distinct(Sequence, .keep_all = TRUE) 

head(pm1088_interpro_distinct_colon02) #no results
```

```r
#NeuS

neuS_interpro_data_colon02 <- read_delim("Interpro_results/colon02_neuS_ID_sequence_for_Interpro.tsv_retrieved_now_results.tsv", 
                                    delim = "\t", escape_double = FALSE, 
                                    col_names = FALSE, trim_ws = TRUE)
head(neuS_interpro_data_colon02)

setnames(neuS_interpro_data_colon02, column_names_interpro)


neuS_interpro_annot_unique_colon02 <- unique(neuS_interpro_data_colon02$annotation)


patterns_neuS <- c("PF07388")

filtered_neuS_interpro_colon02 <- neuS_interpro_data_colon02 %>%
  filter(signature %in% patterns_neuS)

neuS_interpro_distinct_colon02 <- filtered_neuS_interpro_colon02 %>% distinct(Sequence, .keep_all = TRUE)
```

```r
#Lst

Lst_Interpro_data_colon02 <- read_delim("Interpro_results/colon02_lst_ID_sequence_for_Interpro.tsv_retrieved_now_results.tsv", 
                                   delim = "\t", escape_double = FALSE, 
                                   col_names = FALSE, trim_ws = TRUE)
head(Lst_Interpro_data_colon02)

setnames(Lst_Interpro_data_colon02, column_names_interpro)

lst_interpro_annot_unique_colon02 <- unique(Lst_Interpro_data_colon02$annotation)



patterns_Lst <- c("PF07922")

filtered_Lst_interpro_colon02 <- Lst_Interpro_data_colon02 %>%
  filter(signature %in% patterns_Lst)

head(filtered_Lst_interpro_colon02)
lst_filtered_annot_unique_colon02 <- unique(filtered_Lst_interpro_colon02$annotation)

Lst_interpro_distinct_colon02 <- filtered_Lst_interpro_colon02 %>% distinct(Sequence, .keep_all = TRUE)
```

```r
#join all data


#extract GCF ID
CMP_interpro_distinct_colon02$GCF_ID <- str_extract(CMP_interpro_distinct_colon02$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")

#get CMP GCF ID
get_CMP_final_interpro_ID_colon02 <- CMP_interpro_distinct_colon02[,"GCF_ID"]

#NeuS
neuS_interpro_distinct_colon02$GCF_ID <- str_extract(neuS_interpro_distinct_colon02$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")
get_neuS_final_interpro_ID_colon02 <- neuS_interpro_distinct_colon02[,"GCF_ID"]

#NeuS PFAM
neuS_PFAM_interpro_distinct_colon02$GCF_ID <- str_extract(neuS_PFAM_interpro_distinct_colon02$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")
get_neuS_PFAM_final_interpro_ID_colon02 <- neuS_PFAM_interpro_distinct_colon02[,"GCF_ID"]

#Lst 
Lst_interpro_distinct_colon02$GCF_ID <- str_extract(Lst_interpro_distinct_colon02$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")
get_Lst_final_interpro_ID_colon02 <- Lst_interpro_distinct_colon02[,"GCF_ID"]
```

```r
#join all sialyltransferase
all_sialyltrans_ID_colon02 <- rbind(get_neuS_final_interpro_ID_colon02,get_Lst_final_interpro_ID_colon02,
                               get_neuS_PFAM_final_interpro_ID_colon02)

all_sialyltrans_ID_unique_colon02 <-all_sialyltrans_ID_colon02 %>% distinct(GCF_ID)
```

```r
#After filtration, see what are complete pathway that match with CMP
complete_sialylation_interpro_colon02 <- semi_join(get_CMP_final_interpro_ID_colon02, all_sialyltrans_ID_unique_colon02, by="GCF_ID")
```

```r
#Save output
write_tsv(complete_sialylation_interpro_colon02, 
          "Interpro_results/complete_sialylation_interpro_filtration_final_colon02_metagen_update.tsv")
```
