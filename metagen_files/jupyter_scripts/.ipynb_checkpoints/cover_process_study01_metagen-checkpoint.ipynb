```r
#Upload libraries

library(dplyr)
library(readr)
library(tidyr)
library(data.table)
library(stringr)
```

```r
setwd("/mnt/l/Sia_git/metagen_files/Study01_france_cancer/")
```

```r
#setwd to where hmmer coverage output are located.
```

```r
#upload data CMP France metagenomic study 01 colon cancer
View_cmp_FR_cancer <-read.table("output_data/hmmer_out/all_neuA_cover_FR", 
                           header = FALSE, col.names = c("V1", "V2"))

head(View_cmp_FR_cancer)
nrow(View_cmp_FR_cancer)
```

```r
#cover filter

CMP_cover_filter_FR <- View_cmp_FR_cancer %>% filter(View_cmp_FR_cancer$V1 >= 40.00)
head(CMP_cover_filter_FR)
nrow(CMP_cover_filter_FR)
```

```r
# Save result into a file
write_tsv(CMP_cover_filter_FR, "metagen_hmmer_results/CMP_FR_cover_filter_40.tsv")
```
