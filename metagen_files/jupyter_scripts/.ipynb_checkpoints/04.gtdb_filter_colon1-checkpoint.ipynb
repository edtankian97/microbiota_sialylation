```r
# Upload libraries
library(readr)
library(dplyr)
library(stringr)
```

```r
setwd("/mnt/l/colon1_hmmer_backup/")
```

```r
#get unique species

all_gtdbtk_bac120_summary_colon01 <- read_delim("GTDB_results/gtdbtk.bac120.summary.tsv", 
                                            delim = "\t", escape_double = FALSE, 
                                            trim_ws = TRUE)
head(all_gtdbtk_bac120_summary_colon01)
```

```r
# Upload result of checkm filtering

checkm_summary_colon01 <- read_delim("quality_report_bin_filtered_ID.tsv",
                                     delim = "\t", escape_double = FALSE, 
                                     trim_ws = TRUE, col_names = "user_genome")
head(checkm_summary_colon01)
```

```r
# Filter ID based on checkm result
gtdbtk_bac120_summary_colon01 <- semi_join(all_gtdbtk_bac120_summary_colon01, checkm_summary_colon01, by= "user_genome")
nrow(gtdbtk_bac120_summary_colon01)
```

```r
#get species name
gtdbtk_classify_colon01 <- gtdbtk_bac120_summary_colon01[,"classification"]

gtdbtk_species_colon01 <- gtdbtk_classify_colon01 %>%
  mutate(species = str_extract(classification, "(?<=s__).*"))

gtdbtk_species_colon01$species <- trimws(gtdbtk_species_colon01$species)
gtdbtk_species_colon01$species[gtdbtk_species_colon01$species == ""] <- NA

unique(gtdbtk_species_colon01$species)
```

```r
#count species match
table(gtdbtk_species_colon01$species)

count_gtdb_colon01 <- gtdbtk_species_colon01 %>%
  count(species, name = "occurrences")

head(count_gtdb_colon01)
sum(count_gtdb_colon01$occurrences)

unique_gtdbtk_species_colon01 <- gtdbtk_species_colon01 %>% distinct(species, .keep_all = TRUE)

unique_gtdbtk_species_colon01_ID <- unique_gtdbtk_species_colon01[,"species"]
```

```r
write_tsv(unique_gtdbtk_species_colon01_ID, "GTDB_results/unique_gtdbtk_species_colon01_ID.tsv")
```

```r
################################################################################################################
#Division of disease versus Healthy


# Remove everything after the first "-"
gtdbtk_bac120_summary_colon01$ID <- sub("_.*", "", gtdbtk_bac120_summary_colon01$user_genome)

#get ID
gtdbtk_colon01_ID_name <- gtdbtk_bac120_summary_colon01[,"ID"]
```

```r
#save file
write_tsv(gtdbtk_colon01_ID_name, "GTDB_results/gtdbtk_colon01_ERR_ID_name.tsv")
```

```r
#Upload data with health and disease ID

#case
Study_1_cancer_case_Run_ID <- read_tsv("all_cancer_FR_get_ID.tsv")

head(Study_1_cancer_case_Run_ID)
```

```r
#control

Study_1_cancer_control_Run_ID <- read_tsv("france_normal_get_ID.tsv")
head(Study_1_cancer_control_Run_ID)
```

```r
#match and see how many match
colnames(gtdbtk_colon01_ID_name) <- "Run"

#for case
match_colon01_case <- semi_join(gtdbtk_colon01_ID_name,Study_1_cancer_case_Run_ID, by= "Run")
nrow(match_colon01_case)

#get unique IDs
match_colon01_case_unique <- match_colon01_case %>% distinct(Run)
nrow(match_colon01_case_unique)
```

```r
#for control
match_colon01_control <- semi_join(gtdbtk_colon01_ID_name,Study_1_cancer_control_Run_ID, by= "Run")
nrow(match_colon01_control)
#get unique IDs

match_colon01_control_unique <- match_colon01_control %>% distinct(Run)
nrow(match_colon01_control_unique)
```

```r
#Final table 

cat_gtdk_ID <- merge(gtdbtk_bac120_summary_colon01,gtdbtk_species_colon01, by="classification")

cat_gtdk_ID_col <- cat_gtdk_ID[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species")]
head(cat_gtdk_ID)
nrow(cat_gtdk_ID)
```

```r
cat_gtdk_ID_unique <- cat_gtdk_ID%>% distinct(user_genome,.keep_all=TRUE)
nrow(cat_gtdk_ID_unique)
head(cat_gtdk_ID_unique)
```

```r
cat_gtdk_ID_col <- cat_gtdk_ID_unique[,c("classification","closest_genome_reference","closest_genome_ani", "ID", "species", "user_genome")]


cat_colon01_preview <- merge(cat_gtdk_ID_col,count_gtdb_colon01, by= "species")
head(cat_colon01_preview)
```

```r
# handling with control and disease column

match_colon01_control$State <- "Control"
match_colon01_case$State <- "Disease"

# Change column name

colnames(match_colon01_control)[colnames(match_colon01_control) == "Run"] <- "ID"
colnames(match_colon01_case)[colnames(match_colon01_case) == "Run"] <- "ID" 

all_state <- rbind(match_colon01_control,match_colon01_case)

cat_colon01_final <- merge(cat_colon01_preview,all_state, by= "ID")
head(cat_colon01_final)
nrow(cat_colon01_final)
```

```r
cat_colon01_final_end <- cat_colon01_final %>% distinct(user_genome, .keep_all = TRUE)
head(cat_colon01_final_end)
```

```r
#save output
write_tsv(cat_colon01_final_end, "GTDB_results/final_table_gtdb_for_paper_colon01.tsv")
write_tsv(count_gtdb_colon01, "GTDB_results/species_occurrence.tsv")
```

```r
# Classify species
classification <- cat_colon01_final_end %>%
  filter(!is.na(species), !is.na(State)) %>%
  group_by(species) %>%
  summarise(
    Group_classification = case_when(
      n_distinct(State) == 1 & first(State) == "Disease" ~ "Disease",
      n_distinct(State) == 1 & first(State) == "Control" ~ "Control",
      n_distinct(State) > 1 ~ "Control/Disease"
    ),
    .groups = "drop"
  )

# Join  and apply rule for Species == NA
cat_colon01_final_end_df <- cat_colon01_final_end %>%
  left_join(classification, by = "species") %>%
  mutate(
    Group_classification = ifelse(
      is.na(species),
      State,
      Group_classification
    )
  )

head(cat_colon01_final_end_df)
nrow(cat_colon01_final_end_df)
```

```r
write_tsv(cat_colon01_final_end_df,"L:/colon1_hmmer_backup/GTDB_results/gtdb_final_table_with_condition_colon01_corrigido.tsv")
```

```r
cat_colon01_final_end_df_col <- cat_colon01_final_end_df[ ,c("species", "occurrences", "Group_classification")]

cat_colon01_final_end_df_col_unique <- cat_colon01_final_end_df_col %>% distinct(species, .keep_all = TRUE)
head(cat_colon01_final_end_df_col_unique)
nrow(cat_colon01_final_end_df_col_unique)
```

```r
#Save output for paper 
write_tsv(cat_colon01_final_end_df_col_unique, "L:/colon1_hmmer_backup/GTDB_results/count_ocurrence_w_condition_for_paper_colon01.tsv")
```
