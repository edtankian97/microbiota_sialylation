```r
#Upload data
View_neuA_atero01 <-fread("all_CMP_neuA_atero01_all_process_final_process_filtered.tsv",
                          fill =Inf
)[, 1:25]

# set name of header
column_names <-c("target_ID", "accession_1", "tlen", "query_ID", "accession_2", "qlen", "e_value_seq",
                 "score_seq", "bias_seq", "#", "of", "c_value", "i_value", "score_dom", "bias_dom", 
                 "hmm_from", "hmm_to", "ali_from", "ali_to", "env_from", "env_to", "acc", "description",
                 "of_1", "target")

setnames(View_neuA_atero01, column_names)

head(View_neuA_atero01)
nrow(View_neuA_atero01)
```

```r
# see unique IDs for target_ID column
View_neuA_atero01_process <- View_neuA_atero01  %>% distinct(target_ID, .keep_all = TRUE)

#Filter by e-value and bit-score
final_neuA_atero01_process <-View_neuA_atero01_process %>% 
  filter(View_neuA_atero01_process$e_value_seq <= 0.001 & View_neuA_atero01_process$score_seq >= 50.0)

nrow(final_neuA_atero01_process)


#extract ID
final_neuA_atero01_process$GCF_ID <- str_extract(final_neuA_atero01_process$target_ID,
                                                 ".*?\\.faa")

# see unique IDs by GCF_ID column
final_neuA_atero01_process_ID <- final_neuA_atero01_process%>% distinct(GCF_ID, .keep_all = TRUE)

#extract ID
neuA_ID_atero01 <- final_neuA_atero01_process_ID[,"GCF_ID"]
head(neuA_ID_atero01)
nrow(neuA_ID_atero01)
```

```r
#####################################################################################################################################

###########

#hmm_final_process - See what is commom between the files

#CMP versus neuS - thais' database
CMP_plus_neuS_comm_40_complete_atero01  <- intersect(neuA_ID_atero01$GCF_ID, neuS_ID_atero01$GCF_ID)
TRUE_CMP_neuS_thais_data_40_complete_atero01 <- as.data.frame(CMP_plus_neuS_comm_40_complete_atero01)
is.data.frame(TRUE_CMP_neuS_thais_data_40_complete_atero01)
head(TRUE_CMP_neuS_thais_data_40_complete_atero01)
nrow(TRUE_CMP_neuS_thais_data_40_complete_atero01)

#CMP versus Lic3 - thais' database
CMP_plus_lic3_comm_40_complete_atero01  <- intersect(neuA_ID_atero01$GCF_ID, lic3X_ID_atero01$GCF_ID)
TRUE_CMP_lic3_thais_data_40_complete_atero01 <- as.data.frame(CMP_plus_lic3_comm_40_complete_atero01)
is.data.frame(TRUE_CMP_lic3_thais_data_40_complete_atero01)
head(TRUE_CMP_lic3_thais_data_40_complete_atero01)
nrow(TRUE_CMP_lic3_thais_data_40_complete_atero01)

#CMP versus Lst - thais' database
CMP_plus_lst_comm_40_complete_atero01  <- intersect(neuA_ID_atero01$GCF_ID, lst_ID_atero01$GCF_ID)
TRUE_CMP_lst_thais_data_40_complete_atero01 <- as.data.frame(CMP_plus_lst_comm_40_complete_atero01)
is.data.frame(TRUE_CMP_lst_thais_data_40_complete_atero01)
head(TRUE_CMP_lst_thais_data_40_complete_atero01)
nrow(TRUE_CMP_lst_thais_data_40_complete_atero01)


#CMP versus pm0188 - thais' database
CMP_plus_pm0188_comm_40_complete_atero01  <- intersect(neuA_ID_atero01$GCF_ID, pm0188_ID_atero01$GCF_ID)
TRUE_CMP_pm0188_thais_data_40_complete_atero01 <- as.data.frame(CMP_plus_pm0188_comm_40_complete_atero01)
is.data.frame(TRUE_CMP_pm0188_thais_data_40_complete_atero01)
head(TRUE_CMP_pm0188_thais_data_40_complete_atero01)
nrow(TRUE_CMP_pm0188_thais_data_40_complete_atero01)

#CMP versus pm1088 - PFAM database
CMP_plus_PF11477_comm_40_complete_atero01  <- intersect(neuA_ID_atero01$GCF_ID, PF11477_ID_atero01$GCF_ID)
TRUE_CMP_PF11477_thais_data_40_complete_atero01 <- as.data.frame(CMP_plus_PF11477_comm_40_complete_atero01)
is.data.frame(TRUE_CMP_PF11477_thais_data_40_complete_atero01)
head(TRUE_CMP_PF11477_thais_data_40_complete_atero01)
nrow(TRUE_CMP_PF11477_thais_data_40_complete_atero01)

# cst2 PFAM versus CMP
CMP_plus_PF06002_comm_40_complete_atero01  <- intersect(neuA_ID_atero01$GCF_ID, PF06002_ID_atero01$GCF_ID)
TRUE_CMP_PF06002_thais_data_40_complete_atero01 <- as.data.frame(CMP_plus_PF06002_comm_40_complete_atero01)
is.data.frame(TRUE_CMP_PF06002_thais_data_40_complete_atero01)
head(TRUE_CMP_PF06002_thais_data_40_complete_atero01)
nrow(TRUE_CMP_PF06002_thais_data_40_complete_atero01)

#CMP versus neuS - PFAM database
CMP_plus_IPR010866_comm_40_complete_atero01  <- intersect(neuA_ID_atero01$GCF_ID, IPR010866_ID_atero01$GCF_ID)
TRUE_CMP_IPR010866_thais_data_40_complete_atero01 <- as.data.frame(CMP_plus_IPR010866_comm_40_complete_atero01)
is.data.frame(TRUE_CMP_IPR010866_thais_data_40_complete_atero01)
head(TRUE_CMP_IPR010866_thais_data_40_complete_atero01)
nrow(TRUE_CMP_IPR010866_thais_data_40_complete_atero01)
```

```r
###########################################################################################################

#Retrieve IDs to get sequences for Interpro analysis

# rename colnames to match exactly
colnames(TRUE_CMP_IPR010866_thais_data_40_complete_atero01) <- "GCF_ID"
colnames(TRUE_CMP_PF06002_thais_data_40_complete_atero01) <- "GCF_ID"
colnames(TRUE_CMP_PF11477_thais_data_40_complete_atero01) <- "GCF_ID"
colnames(TRUE_CMP_pm0188_thais_data_40_complete_atero01) <- "GCF_ID"
colnames(TRUE_CMP_lst_thais_data_40_complete_atero01) <- "GCF_ID"
colnames(TRUE_CMP_lic3_thais_data_40_complete_atero01) <- "GCF_ID"
colnames(TRUE_CMP_neuS_thais_data_40_complete_atero01) <- "GCF_ID"

all_data_atero01 <- rbind(TRUE_CMP_IPR010866_thais_data_40_complete_atero01,TRUE_CMP_PF06002_thais_data_40_complete_atero01,
                          TRUE_CMP_PF11477_thais_data_40_complete_atero01,TRUE_CMP_pm0188_thais_data_40_complete_atero01,
                          TRUE_CMP_lst_thais_data_40_complete_atero01,TRUE_CMP_lic3_thais_data_40_complete_atero01,
                          TRUE_CMP_neuS_thais_data_40_complete_atero01)

#Sequences with complete sialylation that are not filtered in Interpro
all_data_atero01_unique <-  all_data_atero01 %>% distinct(GCF_ID)

head(all_data_atero01_unique)
nrow(all_data_atero01_unique)
```

```r
#save output
write_tsv(all_data_atero01_unique, "completesialylation_GCF_ID_atero01.tsv")
```

```r
#retrieve sequences

#CMP sialic acid synthase

get_ID_CMP_sequence_atero01 <- semi_join(final_neuA_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
head(get_ID_CMP_sequence_atero01)

#select column
get_ID_CMP_sequence_col_atero01 <- get_ID_CMP_sequence_atero01[,1]


#lst
get_ID_lst_sequence_atero01 <- semi_join(final_lst_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
get_ID_lst_sequence_col_atero01 <- get_ID_lst_sequence_atero01[,1]


#IPR010866
get_ID_IPR010866_sequence_atero01 <- semi_join(final_IPR010866_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
get_ID_IPR010866_sequence_col_atero01 <- get_ID_IPR010866_sequence_atero01[,1]

#PF06002
get_ID_PF06002_sequence_atero01 <- semi_join(final_PF06002_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
get_ID_PF06002_sequence_col_atero01 <- get_ID_PF06002_sequence_atero01[,1]

#PF11477
get_ID_PF11477_sequence_atero01 <- semi_join(final_PF11477_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
get_ID_PF11477_sequence_col_atero01 <- get_ID_PF11477_sequence_atero01[,1]

#pm0188
get_ID_pm0188_sequence_atero01 <- semi_join(final_pm0188_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
get_ID_pm0188_sequence_col_atero01 <- get_ID_pm0188_sequence_atero01[,1]

#lic3
get_ID_lic3_sequence_atero01 <- semi_join(final_lic3X_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
get_ID_lic3_sequence_col_atero01 <- get_ID_lic3_sequence_atero01[,1]

#neuS
get_ID_neuS_sequence_atero01 <- semi_join(final_neuS_atero01_process_ID, all_data_atero01_unique, by="GCF_ID")
get_ID_neuS_sequence_col_atero01 <- get_ID_neuS_sequence_atero01[,1]
```

```r
#save output for Interpro analysis
write_tsv(get_ID_CMP_sequence_col_atero01,"atero01_CMP_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_lst_sequence_col_atero01,"atero01_lst_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_IPR010866_sequence_col_atero01,"atero01_IPR010866_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_neuS_sequence_col_atero01,"atero01_neuS_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_pm0188_sequence_col_atero01,"atero01_pm0188_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_PF11477_sequence_col_atero01,"atero01_PF11477_ID_sequence_for_Interpro.tsv")
```
