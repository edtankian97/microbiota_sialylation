```r
matched_CMP_ID_interpro_colon03 <- read_delim("Interpro_results/colon03_CMP_ID_sequence_for_Interpro_retrieved_now_results.tsv", 
                                              delim = "\t", escape_double = FALSE, 
                                              col_names = FALSE, trim_ws = TRUE)


column_names_interpro <-c("Sequence", "md5", "unkown", "Bank", "signature", "annotation", "align_begin",
                          "align_end", "e-value","Hit", "date", "Interpro_code", "annotation_two", "unkown2", "unkown3")

setnames(matched_CMP_ID_interpro_colon03, column_names_interpro)

unique(matched_CMP_ID_interpro_colon03$annotation)

patterns_of_teste_CMP <- c("TIGR00466","PTHR42866","NF003952", "cd02517", "MF_00057", "TIGR03584") #remove only other nonulosonic citidyltransferase
```

```r
##################################################################################################
#join all data


#extract GCF ID
CMP_interpro_distinct_colon03$GCF_ID <- str_extract(CMP_interpro_distinct_colon03$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")

#get CMP GCF ID
get_CMP_final_interpro_ID_colon03 <- CMP_interpro_distinct_colon03[,"GCF_ID"]


#pm1088
pm1088_interpro_distinct_colon03$GCF_ID <- str_extract(pm1088_interpro_distinct_colon03$Sequence, "^[^_]+_[^_]+_[^_]+")
get_pm1088_final_interpro_ID_colon03 <- pm1088_interpro_distinct_colon03[,"GCF_ID"]


#NeuS
neuS_interpro_distinct_colon03$GCF_ID <- str_extract(neuS_interpro_distinct_colon03$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")
get_neuS_final_interpro_ID_colon03 <- neuS_interpro_distinct_colon03[,"GCF_ID"]

#NeuS PFAM
neuS_PFAM_interpro_distinct_colon03$GCF_ID <- str_extract(neuS_PFAM_interpro_distinct_colon03$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")
get_neuS_PFAM_final_interpro_ID_colon03 <- neuS_PFAM_interpro_distinct_colon03[,"GCF_ID"]

#Lst 
Lst_interpro_distinct_colon03$GCF_ID <- str_extract(Lst_interpro_distinct_colon03$Sequence, "^[^_]+_[^_]+_[^_]+_[^_]+")
get_Lst_final_interpro_ID_colon03 <- Lst_interpro_distinct_colon03[,"GCF_ID"]

#pm1088 - PFAM
pm1088_PFAM_interpro_distinct_colon03$GCF_ID <- str_extract(pm1088_PFAM_interpro_distinct_colon03$Sequence, "^[^_]+_[^_]+_[^_]+")
get_pm1088_PFAM_final_interpro_ID_colon03 <- pm1088_PFAM_interpro_distinct_colon03[,"GCF_ID"]
```

```r
#join all sialyltransferase
all_sialyltrans_ID_colon03 <- rbind(get_neuS_final_interpro_ID_colon03,get_Lst_final_interpro_ID_colon03,
                                    get_neuS_PFAM_final_interpro_ID_colon03,get_pm1088_PFAM_final_interpro_ID_colon03,
                                    get_pm1088_final_interpro_ID_colon03)

all_sialyltrans_ID_unique_colon03 <-all_sialyltrans_ID_colon03 %>% distinct(GCF_ID)
```

```r
#After filtration, see what are complete pathway that match with CMP
complete_sialylation_interpro_colon03 <- semi_join(get_CMP_final_interpro_ID_colon03, all_sialyltrans_ID_unique_colon03, by="GCF_ID")
```
