```r
#Upload data for CMP
matched_CMP_ID_interpro_FR <- read_delim("metagen_files/Study01_france_cancer/output_data/Interpro_results/CMP_ID_sequence_for_Interpro_france.tsv_retrieved_results.tsv", 
                                      delim = "\t", escape_double = FALSE, 
                                      col_names = FALSE, trim_ws = TRUE)
head(matched_CMP_ID_interpro_FR)
```

```r
#rename header
column_names_interpro <-c("Sequence", "md5", "unkown", "Bank", "signature", "annotation", "align_begin",
                          "align_end", "e-value","Hit", "date", "Interpro_code", "annotation_two", "unkown2", "unkown3")

setnames(matched_CMP_ID_interpro_FR, column_names_interpro)

unique(matched_CMP_ID_interpro_FR$annotation)

patterns_of_teste_CMP <- c("TIGR00466","PTHR42866","NF003952", "cd02517", "MF_00057", "TIGR03584") #remove only other nonulosonic citidyltransferase
```

# Filter sequences with annotation to other nonusolosonic 

```r
# 1. Find all Sequence IDs containing any of the patterns
bad_ids <- matched_CMP_ID_interpro_FR %>%
  filter(grepl(paste(patterns_of_teste_CMP, collapse = "|"), signature)) %>%
  pull(Sequence) %>%
  unique()

# 2. Remove those IDs entirely
filtered_CMP_interpro_rm_FR <- matched_CMP_ID_interpro_FR %>%
  filter(!(Sequence %in% bad_ids))
```

```r
patterns <- c("PF02348","SSF53448","G3DSA:3.90.550.10","PTHR21485","cd02513")

filtered_CMP_interpro_FR <- filtered_CMP_interpro_rm_FR %>%
  filter(signature %in% patterns)

View(filtered_CMP_interpro_FR)
CMP_interpro_distinct_FR <- filtered_CMP_interpro_FR%>% distinct(Sequence, .keep_all = TRUE)
```

# Join all data

```r
#extract GCF ID
CMP_interpro_distinct_FR$GCF_ID <- str_extract(CMP_interpro_distinct_FR$Sequence, "^[^_]+_[^_]+_[^_]+")

#get CMP GCF ID
get_CMP_final_interpro_ID_FR <- CMP_interpro_distinct_FR[,"GCF_ID"]


#pm1088
pm1088_interpro_distinct_FR$GCF_ID <- str_extract(pm1088_interpro_distinct_FR$Sequence, "^[^_]+_[^_]+_[^_]+")
get_pm1088_final_interpro_ID_FR <- pm1088_interpro_distinct_FR[,"GCF_ID"]


#NeuS
neuS_interpro_distinct_FR$GCF_ID <- str_extract(neuS_interpro_distinct_FR$Sequence, "^[^_]+_[^_]+_[^_]+")
get_neuS_final_interpro_ID_FR <- neuS_interpro_distinct_FR[,"GCF_ID"]

#NeuS PFAM
neuS_PFAM_interpro_distinct_FR$GCF_ID <- str_extract(neuS_PFAM_interpro_distinct_FR$Sequence, "^[^_]+_[^_]+_[^_]+")
get_neuS_PFAM_final_interpro_ID_FR <- neuS_PFAM_interpro_distinct_FR[,"GCF_ID"]

#Lst 
Lst_interpro_distinct_FR$GCF_ID <- str_extract(Lst_interpro_distinct_FR$Sequence, "^[^_]+_[^_]+_[^_]+")
get_Lst_final_interpro_ID_FR <- Lst_interpro_distinct_FR[,"GCF_ID"]
```

```r
#join all sialyltransferase
all_sialyltrans_ID_FR <- rbind(get_pm1088_final_interpro_ID_FR, get_neuS_final_interpro_ID_FR,get_Lst_final_interpro_ID_FR,
                            get_neuS_PFAM_final_interpro_ID_FR)

all_sialyltrans_ID_unique_FR <-all_sialyltrans_ID_FR %>% distinct(GCF_ID)
```

```r
#After filtration, see what are complete pathway that match with CMP
complete_sialylation_interpro_FR <- semi_join(get_CMP_final_interpro_ID_FR, all_sialyltrans_ID_unique_FR, by="GCF_ID")
```

```r
write_tsv(complete_sialylation_interpro_FR, 
          "metagen_files/Study01_france_cancer/output_data/Interpro_results/complete_sialylation_interpro_filtration_final_FR_metagen_update.tsv")
```

```r
#get bins
get_bins_FR <- semi_join(CMP_interpro_distinct_FR, complete_sialylation_interpro_FR, by="GCF_ID")
head(get_bins_FR)
```

```r
get_bins_FR_name <- get_bins_FR[,1]
get_bins_FR_name <- get_bins_FR_name %>%
  mutate(Sequence = Sequence %>%
           str_remove("^PROKKA_") %>%
           str_replace("\\.faa.*$", ".fa"))
```
