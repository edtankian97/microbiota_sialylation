```r
###########################################################################################################

#Retrieve IDs to get sequences for Interpro analysis

# rename colnames to match exactly
colnames(TRUE_CMP_IPR010866_thais_data_40_complete_colon03) <- "GCF_ID"
colnames(TRUE_CMP_PF06002_thais_data_40_complete_colon03) <- "GCF_ID"
colnames(TRUE_CMP_PF11477_thais_data_40_complete_colon03) <- "GCF_ID"
colnames(TRUE_CMP_pm0188_thais_data_40_complete_colon03) <- "GCF_ID"
colnames(TRUE_CMP_lst_thais_data_40_complete_colon03) <- "GCF_ID"
colnames(TRUE_CMP_lic3_thais_data_40_complete_colon03) <- "GCF_ID"
colnames(TRUE_CMP_neuS_thais_data_40_complete_colon03) <- "GCF_ID"

all_data_colon03 <- rbind(TRUE_CMP_IPR010866_thais_data_40_complete_colon03,TRUE_CMP_PF06002_thais_data_40_complete_colon03,
                          TRUE_CMP_PF11477_thais_data_40_complete_colon03,TRUE_CMP_pm0188_thais_data_40_complete_colon03,
                          TRUE_CMP_lst_thais_data_40_complete_colon03,TRUE_CMP_lic3_thais_data_40_complete_colon03,
                          TRUE_CMP_neuS_thais_data_40_complete_colon03)
```

```r
#Sequences with complete sialylation that are not filtered in Interpro
all_data_colon03_unique <-  all_data_colon03 %>% distinct(GCF_ID)

head(all_data_colon03_unique)
nrow(all_data_colon03_unique)
```

```r
#save output
write_tsv(all_data_colon03_unique, "completesialylation_GCF_ID_colon03.tsv")
```
