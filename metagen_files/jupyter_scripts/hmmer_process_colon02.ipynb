```r
#Get data from CMP-NeuA

# Data without a header
View_CMP_colon_02 <- fread("Study02_colon_cancer/output_data/all_CMP_neuA_colon02_all_process_final_process_filtered.tsv", fill=25)


column_names <-c("target_ID", "accession_1", "tlen", "query_ID", "accession_2", "qlen", "e_value_seq",
                 "score_seq", "bias_seq", "#", "of", "c_value", "i_value", "score_dom", "bias_dom", 
                 "hmm_from", "hmm_to", "ali_from", "ali_to", "env_from", "env_to", "acc", "description",
                 "of_1", "target")

setnames(View_CMP_colon_02, column_names)

head(View_CMP_colon_02)
```

```r
#Retrieve IDs to get sequences for Interpro analysis

# rename colnames to match exactly
colnames(TRUE_CMP_neuS_thais_data_40_complete) <- "target_ID"
colnames(TRUE_CMP_Lic3_thais_data_40_complete) <- "target_ID"
colnames(TRUE_CMP_IPR010866_data_40_complete) <- "target_ID"
colnames(TRUE_CMP_PF11477_data_40_complete) <- "target_ID"
colnames(TRUE_CMP_lst_data_40_complete) <- "target_ID"
colnames(TRUE_CMP_pm0188_data_40_complete) <- "target_ID"
colnames(TRUE_CMP_PF06002_data_40_complete) <- "target_ID"
```

```r
#join all
all_data_colon02 <- rbind(TRUE_CMP_neuS_thais_data_40_complete,TRUE_CMP_Lic3_thais_data_40_complete,TRUE_CMP_IPR010866_data_40_complete,
                          TRUE_CMP_PF11477_data_40_complete,TRUE_CMP_lst_data_40_complete,TRUE_CMP_pm0188_data_40_complete,
                          TRUE_CMP_PF06002_data_40_complete)
head(all_data_colon02)
nrow(all_data_colon02)
```

```r
#Sequences with complete sialylation that are not filtered in Interpro
all_data_colon02_unique <-  all_data_colon02 %>% distinct(target_ID)
```

```r
#save output
write_tsv(all_data_colon02_unique, "Study02_colon_cancer/completesialylation_GCF_ID_colon02.tsv")
```

```r
#retrieve sequences for Interpro that returned with any hit after filter - lst, IPR010866, neuS and pm0188

#CMP
get_ID_CMP_sequence <- semi_join(final_CMP_colon02_process_ID, all_data_colon02_unique, by="target_ID")
get_ID_CMP_sequence_col <- get_ID_CMP_sequence[,"file_ID"]

#lst
get_ID_lst_sequence <- semi_join(final_lst_colon02_process_ID, all_data_colon02_unique, by="target_ID")
get_ID_lst_sequence_col <- get_ID_lst_sequence[,"file_ID"]

#IPR010866
get_ID_IPR010866_sequence <- semi_join(final_IPR010866_colon02_process_ID, all_data_colon02_unique, by="target_ID")
get_ID_IPR010866_sequence_col <- get_ID_IPR010866_sequence[,"file_ID"]

#neuS
get_ID_neuS_sequence <- semi_join(final_neuS_colon02_process_ID, all_data_colon02_unique, by="target_ID")
get_ID_neuS_sequence_col <- get_ID_neuS_sequence[,"file_ID"]

#pm0188
get_ID_pm0188_sequence <- semi_join(final_pm0188_colon02_process_ID, all_data_colon02_unique, by="target_ID")
get_ID_pm0188_sequence_col <- get_ID_pm0188_sequence[,"file_ID"]
```

```r
#save output
write_tsv(get_ID_CMP_sequence_col,"Study02_colon_cancer/colon02_CMP_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_lst_sequence_col,"Study02_colon_cancer/colon02_lst_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_IPR010866_sequence_col,"Study02_colon_cancer/colon02_IPR010866_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_neuS_sequence_col,"Study02_colon_cancer/colon02_neuS_ID_sequence_for_Interpro.tsv")
write_tsv(get_ID_pm0188_sequence_col,"Study02_colon_cancer/colon02_pm0188_ID_sequence_for_Interpro.tsv")
```
