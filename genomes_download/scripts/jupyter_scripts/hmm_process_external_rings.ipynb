# Part 01 - Coverage filtration

```r
#Library 
library(dplyr)
library(readr)
library(tidyr)
library(data.table)
library(stringr)
```

```r
# Set the working directory to the parent folder of microbiota_sialylation/genomes_download/
# Example:
# setwd("~")
```

```r
#Upload data
#KpsM
cover_kpsM_ncbi <-read.table("plots_data/hmmer_out/all_kpsM_coverage.tsv", 
                                       header = FALSE, col.names = c("V1", "V2"))
head(cover_kpsM_ncbi)
nrow(cover_kpsM_ncbi)
#KpsT
cover_kpsT_ncbi <-read.table("plots_data/hmmer_out/all_kpsT_coverage.tsv", 
                             header = FALSE, col.names = c("V1", "V2"))

head(cover_kpsT_ncbi)
nrow(cover_kpsT_ncbi)
#KpsD
cover_kpsD_ncbi <-read.table("plots_data/hmmer_out/all_KpsD_coverage.tsv", 
                             header = FALSE, col.names = c("V1", "V2"))

head(cover_kpsD_ncbi)
nrow(cover_kpsD_ncbi)

#neuD
cover_neuD_ncbi <-read.table("plots_data/hmmer_out/all_neuD_coverage.tsv", 
                             header = FALSE, col.names = c("V1", "V2"))
head(cover_neuD_ncbi)
nrow(cover_neuD_ncbi)

#neuO
cover_neuO_ncbi <-read.table("plots_data/hmmer_out/all_neuO_coverage.tsv", 
                             header = FALSE, col.names = c("V1", "V2"))
head(cover_neuO_ncbi)
nrow(cover_neuO_ncbi)
```

```r
#Filter by coverage
#KpsM
cover_kpsM_ncbi_filter <- cover_kpsM_ncbi %>% filter(cover_kpsM_ncbi$V1 >= 40.00)
nrow(cover_kpsM_ncbi_filter)

#KpsT
cover_kpsT_ncbi_filter <- cover_kpsT_ncbi %>% filter(cover_kpsT_ncbi$V1 >= 40.00)
nrow(cover_kpsT_ncbi_filter)

#KpsD
cover_kpsD_ncbi_filter <- cover_kpsD_ncbi %>% filter(cover_kpsD_ncbi$V1 >= 40.00)
nrow(cover_kpsD_ncbi_filter)

#neuO
cover_neuO_ncbi_filter <- cover_neuO_ncbi %>% filter(cover_neuO_ncbi$V1 >= 40.00)
nrow(cover_neuO_ncbi_filter)

#neuD
cover_neuD_ncbi_filter <- cover_neuD_ncbi %>% filter(cover_neuD_ncbi$V1 >= 40.00)
nrow(cover_neuO_ncbi_filter)
```

```r
#save files
write_tsv(cover_kpsM_ncbi_filter, "Protein_database/external_rings_models/external_rings_output/cover_kpsM_ncbi_filter.tsv")
write_tsv(cover_kpsT_ncbi_filter, "Protein_database/external_rings_models/external_rings_output/cover_kpsT_ncbi_filter.tsv")
write_tsv(cover_kpsD_ncbi_filter, "Protein_database/external_rings_models/external_rings_output/cover_kpsD_ncbi_filter.tsv")
write_tsv(cover_neuD_ncbi_filter, "Protein_database/external_rings_models/external_rings_output/cover_neuD_ncbi_filter.tsv")
write_tsv(cover_neuO_ncbi_filter, "Protein_database/external_rings_models/external_rings_output/cover_neuO_ncbi_filter.tsv")
```
