```r
#Upload libraries
library(readr)
library(dplyr)
library(ggplot2)
library(ggrepel)
library(stringr)
```

```r
#Upload data
accession_list_1 <-  read_delim("microbiota_sialylation/genomes_download/plots_data/comm_complete_genomes_dataset_fields.tsv", 
                                 delim = "\t", escape_double = FALSE, 
                                 trim_ws = TRUE)
head(accession_list_1)
colnames(accession_list_1)
```

```r
#select columm that I want
select_data <- accession_list_1%>% select('Assembly Accession', 'Assembly BioSample Host', 'Organism Taxonomic ID')

#ascending format by Taxonomic ID
select_data_asc <- select_data %>% arrange('Organism Taxonomic ID')
```

```r
#Upload taxonomy info
taxonomy_summary <- read_delim("microbiota_sialylation/genomes_download/plots_data/taxonomy_summary.tsv", 
                               delim = "\t", escape_double = FALSE, 
                               trim_ws = TRUE)
head(taxonomy_summary)
```

```r
#let in ascending format by Taxonomic ID
select_taxo_asc <- taxonomy_summary %>% arrange('Query')

#rename data to join after
colnames(select_taxo_asc)[1] <- "Organism Taxonomic ID"
```

```r
#join data
missed_data <- left_join(select_data_asc, select_taxo_asc, by = 'Organism Taxonomic ID')
write_tsv(missed_data, "host_data_comm_sia.tsv")
```

```r
#Preparing data of species to the plot
species_final <-  missed_data %>%
  mutate(first_two_words = word(`Species name`, 1, 2))%>%
  group_by(first_two_words) %>%             # Group by 'region'
  mutate(species_count = n()) %>%   # Count occurrences and add to new column
  ungroup()                        # Remove grouping



species_final <- species_final%>%select(first_two_words, 'species_count')



species_final_unique <- species_final %>%
  distinct(first_two_words, .keep_all = TRUE)


head(species_final_unique)
```

```r
species_final_unique <- species_final_unique %>%
  mutate(perc_species = species_count / sum(species_count) * 100) %>%
  filter(perc_species >= 1.80)

perc_species_round <-round(species_final_unique$perc_species, digits = 2)
```

```r
# Create a pie chart
f<-ggplot(species_final_unique, aes(x = "", y = perc_species_round, fill = first_two_words)) +
  geom_bar(stat = "identity", width = 1, color = "black") +
  coord_polar(theta = "y") +
  theme_void() + # Removes axes and background 
  theme(plot.title = element_text(hjust = 0.5, size = 16, face = "bold", family = "Times"),
        legend.text=element_text(size=12, family  = "Times", face = "italic"),
        legend.title=element_text(size=14, face= "bold", family = "Times")) +
  labs(title = "Species Distribution (%)", fill = "Species Name") +
  geom_label_repel(aes(x = 1.7, label = perc_species_round),
                   color = "white",
                   position = position_stack(vjust = 0.5),
                   show.legend = FALSE,
                   box.padding = 0.2,  # Space around label
                   segment.curvature = 0.2,
                   segment.angle = 20,
                   segment.ncp = 3,
                   segment.color = "grey") +
  scale_fill_viridis_d()

options(repr.plot.width = 14, repr.plot.height = 7)
print(f)
```
