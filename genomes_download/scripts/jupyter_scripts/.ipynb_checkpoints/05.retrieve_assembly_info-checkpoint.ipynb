# Genome info for complete genomes with sialylation pathway

```r
# Set the working directory to the parent folder of genomes_download/
# Example:
# setwd("~/microbiota_sialylation")
```

```r
#Upload libraries
library(ggplot2)
library(readr)
library(ggpubr)
```

```r
#upload data
complete_genomes_dataset <- read_delim("genomes_download/plots_data/accession_complete_fields.tsv", 
                                                   delim = "\t", escape_double = FALSE, 
                                                   trim_ws = TRUE)
head(complete_genomes_dataset)
colnames(complete_genomes_dataset)
```

## Genome length 

```r
#Genome size
genome_complete_sia_phylo <-ggplot(complete_genomes_dataset, aes(x= complete_genomes_dataset$`Assembly Stats Total Sequence Length`)) + 
  geom_histogram(color="black", fill="#7bccc4") +  
  theme(axis.text.x=element_text(angle = 0, size = 22, colour = "black", family = "Times"),
        axis.text.y = element_text(angle = 0, size = 22, colour = "black", family = "Times"),
        axis.text = element_text(size = 22, colour = "black", family = "Times"),
        axis.title=element_text(size=24,face="bold"),
        panel.background = element_blank(),
        axis.line = element_line(colour = "black"), plot.margin = unit(c(3,10,3,3), units = "mm")) + labs(y= "Count", x = "Genome size (bp)") +
  scale_y_continuous(expand = expansion())

options(repr.plot.width = 14, repr.plot.height = 7)
print(genome_complete_sia_phylo)

median(complete_genomes_dataset$`Assembly Stats Total Sequence Length`) #check median of sequence length
```

## G+C content

```r
# G+C content
gc_complete_sia_phylo <-ggplot(complete_genomes_dataset, aes(x=complete_genomes_dataset$`Assembly Stats GC Percent`)) + 
  geom_histogram(color="black", fill="#43a2ca") +  
  theme(axis.text.x=element_text(angle = 0, size = 22, colour = "black", family = "Times"),
        axis.text.y = element_text(angle = 0, size = 2, colour = "black", family = "Times"),
        axis.text = element_text(size = 22, colour = "black", family = "Times"),
        axis.title=element_text(size=24,face="bold"),
        panel.background = element_blank(),
        axis.line = element_line(colour = "black"), plot.margin = unit(c(3,5,3,10), units = "mm")) + labs(y= "Count", x = "GC content (%)") +
  scale_y_continuous(expand = expansion())

options(repr.plot.width = 14, repr.plot.height = 7)
print(gc_complete_sia_phylo)

median(complete_genomes_dataset$`Assembly Stats GC Percent`)
```

## CDS

```r
#CDS
CDS_complete_sia_phylo <-ggplot(complete_genomes_dataset, aes(x=complete_genomes_dataset$`Annotation Count Gene Protein-coding`)) + 
  geom_histogram(color="black", fill="#0868ac") +  
  theme(axis.text.x=element_text(angle = 0, size = 22, colour = "black", family = "Times"),
        axis.text.y = element_text(angle = 0, size = 22, colour = "black", family = "Times"),
        axis.text = element_text(size = 22, colour = "black", family = "Times"),
        axis.title=element_text(size=24,face="bold"),
        panel.background = element_blank(),
        axis.line = element_line(colour = "black"), plot.margin = unit(c(3,5,3,10), units = "mm")) +
  labs(y= "Count", x = "CDS") +
  scale_y_continuous(expand = expansion())

print(CDS_complete_sia_phylo)

median(complete_genomes_dataset$`Annotation Count Gene Protein-coding`)
```

## Join plots

```r
options(repr.plot.width = 14, repr.plot.height = 7)
ggarrange(genome_complete_sia_phylo,CDS_complete_sia_phylo, gc_complete_sia_phylo, nrow = 1, ncol = 3)
```
