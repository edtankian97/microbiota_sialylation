```r
#Upload data for CMP
matched_CMP_ID_interpro <- read_delim("genomes_download/plots_data/Interpro_results/CMP_ID_sequence_for_Interpro.tsv_retrieved_now_results.tsv", 
                                                       delim = "\t", escape_double = FALSE, 
                                                       col_names = FALSE, trim_ws = TRUE)
head(matched_CMP_ID_interpro)
```

```r
#rename header
column_names_interpro <-c("Sequence", "md5", "unkown", "Bank", "signature", "annotation", "align_begin",
                 "align_end", "e-value","Hit", "date", "Interpro_code", "annotation_two", "unkown2", "unkown3")

setnames(matched_CMP_ID_interpro, column_names_interpro)

unique(matched_CMP_ID_interpro$annotation)

patterns_of_teste_CMP <- c("TIGR00466","PTHR42866","NF003952", "cd02517", "MF_00057", "TIGR03584") #remove only other nonulosonic citidyltransferase
```

# Filter sequences with annotation to other nonusolosonic 

```r
# 1. Find all Sequence IDs containing any of the patterns
bad_ids <- matched_CMP_ID_interpro %>%
  filter(grepl(paste(patterns_of_teste_CMP, collapse = "|"), signature)) %>%
  pull(Sequence) %>%
  unique()

# 2. Remove those IDs entirely
filtered_CMP_interpro_rm <- matched_CMP_ID_interpro %>%
  filter(!(Sequence %in% bad_ids))
```

```r
#get real names in annotation

patterns <- c("PF02348","SSF53448","G3DSA:3.90.550.10","PTHR21485","cd02513")

filtered_CMP_interpro <- filtered_CMP_interpro_rm %>%
  filter(signature %in% patterns)

head(filtered_CMP_interpro)
  CMP_interpro_distinct <- filtered_CMP_interpro%>% distinct(Sequence, .keep_all = TRUE)
```

# For Cst Pfam

```r
#cst PFAM

cst_PFAM_Interpro_data <- read_delim("genomes_download/plots_data/Interpro_results/cst_ID_PFAM_sequence_for_Interpro.tsv_retrieved_now_results.tsv", 
                                 delim = "\t", escape_double = FALSE, 
                                 col_names = FALSE, trim_ws = TRUE)
head(cst_PFAM_Interpro_data)
```

```r
setnames(cst_PFAM_Interpro_data, column_names_interpro) #rename header

cst_PFAM_Interpro_annot_unique <- unique(cst_PFAM_Interpro_data$annotation)
cst_PFAM_Interpro_annot_unique #see unique signatures' names

patterns_cst <- c("SSF102414","G3DSA:3.90.1480.10","PF06002") #get ID of signatures

filtered_cst_interpro <- cst_PFAM_Interpro_data %>%
  filter(signature %in% patterns_cst) #filter sequences with cst signatures

cst_PFAM_interpro_distinct <- filtered_cst_interpro %>% distinct(Sequence, .keep_all = TRUE) #remove duplicates
```

```r
#join all data

#extract GCF ID
CMP_interpro_distinct$GCF_ID <- str_extract(CMP_interpro_distinct$Sequence, "^[^_]+_[^_]+")

#get CMP GCF ID
get_CMP_final_interpro_ID <- CMP_interpro_distinct[,"GCF_ID"]

#get ID for all sialyltransferases
#Lic3
Lic3_interpro_distinct$GCF_ID <- str_extract(Lic3_interpro_distinct$Sequence, "^[^_]+_[^_]+")
get_Lic3_final_interpro_ID <- Lic3_interpro_distinct[,"GCF_ID"]

#pm1088
pm1088_interpro_distinct$GCF_ID <- str_extract(pm1088_interpro_distinct$Sequence, "^[^_]+_[^_]+")
get_pm1088_final_interpro_ID <- pm1088_interpro_distinct[,"GCF_ID"]

#pm1088 PFAM
pm1088_PFAM_interpro_distinct$GCF_ID <- str_extract(pm1088_PFAM_interpro_distinct$Sequence, "^[^_]+_[^_]+")
get_pm1088_PFAM_final_interpro_ID <- pm1088_PFAM_interpro_distinct[,"GCF_ID"]

#NeuS
neuS_interpro_distinct$GCF_ID <- str_extract(neuS_interpro_distinct$Sequence, "^[^_]+_[^_]+")
get_neuS_final_interpro_ID <- neuS_interpro_distinct[,"GCF_ID"]

#NeuS PFAM
neuS_PFAM_interpro_distinct$GCF_ID <- str_extract(neuS_PFAM_interpro_distinct$Sequence, "^[^_]+_[^_]+")
get_neuS_PFAM_final_interpro_ID <- neuS_PFAM_interpro_distinct[,"GCF_ID"]

#Lst 
Lst_interpro_distinct$GCF_ID <- str_extract(Lst_interpro_distinct$Sequence, "^[^_]+_[^_]+")
get_Lst_final_interpro_ID <- Lst_interpro_distinct[,"GCF_ID"]

#cst PFAM
cst_PFAM_interpro_distinct$GCF_ID <- str_extract(cst_PFAM_interpro_distinct$Sequence, "^[^_]+_[^_]+")
get_cst_PFAM_final_interpro_ID <- cst_PFAM_interpro_distinct[,"GCF_ID"]
```

# see what have sialylation complete pathway

```r
#join all sialyltransferase
all_sialyltrans_ID <- rbind(get_Lic3_final_interpro_ID,get_pm1088_final_interpro_ID, get_neuS_final_interpro_ID,get_Lst_final_interpro_ID,
                            get_neuS_PFAM_final_interpro_ID, get_cst_PFAM_final_interpro_ID, get_pm1088_PFAM_final_interpro_ID)

all_sialyltrans_ID_unique <-all_sialyltrans_ID %>% distinct(GCF_ID)

#After filtration, see what are complete pathway that match with CMP
complete_sialylation_interpro <- semi_join(get_CMP_final_interpro_ID, all_sialyltrans_ID_unique, by="GCF_ID")
```

```r
#save result to an output
write_tsv(complete_sialylation_interpro, 
          "genomes_download/plots_data/complete_sialylation_interpro_filtration_final.tsv") 
```

# see how many are from Polysialylation pathway

```r
#Polysialylation
get_neuS_PFAM <- semi_join(complete_sialylation_interpro,get_neuS_PFAM_final_interpro_ID, by="GCF_ID")
get_neuS <- semi_join(complete_sialylation_interpro, get_neuS_final_interpro_ID, by="GCF_ID")

all_poly <- rbind(get_neuS_PFAM,get_neuS)
all_poly_unique <- distinct(all_poly)
head(all_poly_unique)
nrow(all_poly_unique)
polysia_path_count <- nrow(all_poly_unique)
```

# see how many are from Sialylation pathway

```r
get_pm0188<- semi_join(complete_sialylation_interpro,pm1088_interpro_distinct, by="GCF_ID")
get_Lic3 <- semi_join(complete_sialylation_interpro, Lic3_interpro_distinct, by="GCF_ID")
get_pm0188_PFAM <- semi_join(complete_sialylation_interpro, pm1088_PFAM_interpro_distinct, by="GCF_ID")
get_cst_PFAM <- semi_join(complete_sialylation_interpro, cst_PFAM_interpro_distinct, by="GCF_ID")
get_lst <- semi_join(complete_sialylation_interpro,Lst_interpro_distinct, by="GCF_ID")

all_sialyl_ID_together <- rbind(get_lst,get_pm0188, get_Lic3,get_pm0188_PFAM,get_cst_PFAM)
all_sialyl_ID_together_unique <-all_sialyl_ID_together %>% distinct(GCF_ID)
head(all_sialyl_ID_together_unique)
nrow(all_sialyl_ID_together_unique)
sia_path_count <- nrow(all_sialyl_ID_together_unique)
```

# See how many has both pathway

```r
#see what have both
join_all_path <-rbind(all_sialyl_ID_together_unique, all_poly_unique)
duplicated_rows <- join_all_path[duplicated(join_all_path), ]#duplicate
nrow(duplicated_rows)
both_path <- nrow(duplicated_rows)
```

```r
#prepare data for Venn diagram
sia_venn <- sia_path_count + both_path
polysia_venn <- polysia_path_count + both_path
```

#  Create barplot for sialylation pathway

```r
# create pairwi4se Venn diagram
draw.pairwise.venn(area1=sia_venn, area2=polysia_venn,cross.area=both_path,
                   category=c("LPS/LOS Sialylation","Polysialylation"),fill=c("#f768a1","#ae017e"),
cex = 2,        # number labels size
cat.cex = 2,    # category labels size
# font family and face for the number labels
fontfamily = "Times",
fontface = "bold",
# font family and face for the category labels
cat.fontfamily = "Times",
cat.fontface = "bold",
cat.pos = c(-10, 10),

# distance from the circles (float numbers): increase to move farther away
cat.dist = c(0.06, 0.06),
)
```

# Create barplot for nonulosonic family count

```r
#get GCF_ID that match with CMP-KDO-synthetase
teste <- matched_CMP_ID_interpro %>%
  filter(signature %in% c("TIGR00466", "PTHR42866", "NF003952", 
                          "cd02517", "MF_00057"))

teste$GCF_ID <- str_extract(teste$Sequence, "^[^_]+_[^_]+") #extract GCF_ID 

teste_unique_ID <- teste %>% distinct(GCF_ID, .keep_all = TRUE) 
teste_ID <- teste[,1]

#number of KDO-synthetase sequences
nrow(teste_unique_ID)
count_teste_unique_ID_KDO <- nrow(teste_unique_ID)

#get GCF_ID that match with CMP-Pse-synthetase
teste_Pse <- matched_CMP_ID_interpro %>%
  filter(signature %in% c("TIGR03584"))

teste_Pse$GCF_ID <- str_extract(teste_Pse$Sequence, "^[^_]+_[^_]+") #extract GCF_ID 

teste_Pse_unique_ID <- teste_Pse %>% distinct(GCF_ID, .keep_all = TRUE) 
teste_ID <- teste[,1]
nrow(teste_Pse_unique_ID)

#number of Pse synthetase sequences 
count_teste_unique_ID_Pse <- nrow(teste_Pse_unique_ID)

#number of CMP-NeuA synthetase sequences
nrow(get_CMP_final_interpro_ID)
count_teste_unique_ID_CMP <- nrow(get_CMP_final_interpro_ID)
```

```r
barplot <- data.frame(Protein=c("NeuA + LegF", "KDO-synthetase","Pse-synthetase"),
           Counts=c(count_teste_unique_ID_CMP,count_teste_unique_ID_KDO,count_teste_unique_ID_Pse))


one <- ggplot(barplot, aes(x = reorder(Protein, -Counts), y = Counts, fill = Protein)) + 
  geom_bar(stat = "identity") + 
  geom_text(aes(label = Counts),
            vjust = -0.5, size = 7, family = "Times") +
  scale_fill_manual(values=c("#ffeda0", "#feb24c", "#f03b20")) +
  theme(axis.text.x= element_blank(),
        axis.text.y = element_text(angle = 0, size = 22, colour = "black", family = "Times"),
        axis.text = element_text(size = 15, colour = "black", family = "Times"),
        axis.title.x = element_blank(),
        axis.title.y = element_text(size=24,face="bold"),
        panel.background = element_blank(),
        axis.line = element_line(colour = "black"),
        legend.text = element_text(size = 22,colour = "black", family = "Times"),
        legend.title = element_text(size = 24,colour = "black", family = "Times", face = "bold"),
        legend.position.inside = c(0.7,0.7)) +
  scale_y_continuous(expand = expansion(), limits = c(0,3100), breaks = c(0,1000,2000,3000))

one
```
