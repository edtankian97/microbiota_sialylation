```r
#VFDB - virulence factors

#select first columm
vfdb_itol_represent <- vfdb_70id_60cov[,1]
colnames(vfdb_itol_represent) <- "Sequence"

#take fisrt hit for each ID
vfdb_itol_represent <- distinct(vfdb_itol_represent)

#Extract ID and some part put in another columm
df_vfdb <- vfdb_itol_represent %>%
  mutate(GCF_ID = str_extract(Sequence, "^[^_]*_[^_]*"))

#add _protein to each line 
df_vfdb$Part1 <- paste0(df_vfdb$GCF_ID, "_protein")

#see what is different between datasets
diff_representative_vfdb <- setdiff(all_GCF_files %>% select(GCF_ID), df_vfdb %>% select(GCF_ID))
head(diff_representative_vfdb)


#Atributte color and attach string "_protein" to IDs which were not identified any protein
diff_representative_vfdb$level <- "#DC143C"

diff_representative_vfdb$GCF_ID <- paste0(diff_representative_vfdb$GCF_ID, "_protein")

write_tsv(diff_representative_vfdb,"genomes_download/plots_data/itol/no_virulence_sialylation.tsv")

#Atributte color and attach string "_protein" to ID's identified with protein
df_vfdb$level <- "#50C878"
df_vfdb1 <- df_vfdb[ , c(-1,-2)]

colnames(df_vfdb1) <- c("GCF_ID", "level")
write_tsv(df_vfdb,"genomes_download/plots_data/itol/virulence_sialylation.tsv")
#Join the data
data_vfdb_itol_represent <- rbind(df_vfdb1, diff_representative_vfdb)
head(data_vfdb_itol_represent)

#save results in a file
write_tsv(data_vfdb_itol_represent, "genomes_download/plots_data/itol/vfdb_itol_represent_EANS.tsv")
```

```r
#label each node with specie name

#Upload data from host_distribution script
host_data_comm_sia_species <- read_delim("genomes_download/plots_data/itol/host_data_comm_sia.tsv", 
                                 delim = "\t", escape_double = FALSE, 
                                 trim_ws = TRUE)
head(host_data_comm_sia_species)

#select colummns
select_ID <- host_data_comm_sia_species%>% select(`Assembly Accession`, `Tax name`)
select_ID_name <-  select_ID  %>%
  mutate(Species_name = word(`Tax name`, 1, 2))

select_ID_name <- select_ID_name[,c(1,3)]
select_ID_name$`Assembly Accession` <- paste0(select_ID_name$`Assembly Accession`, "_protein")


#save file
write_tsv(select_ID_name, "genomes_download/plots_data/itol/representative_labels_itol.tsv")



#label for phylum

host_data_comm_sia <- read_delim("genomes_download/plots_data/itol/host_data_comm_sia.tsv", 
                                 delim = "\t", escape_double = FALSE, 
                                 trim_ws = TRUE)
View(host_data_comm_sia)



select_phylum <- host_data_comm_sia%>% select('Assembly Accession', 'Phylum name')

unique(select_phylum$`Phylum name`)
select_phylum$'Assembly Accession' <- paste0(select_phylum$'Assembly Accession', "_protein")
       
       select_phylum$label_type <- "label_background"
```
