```r
#Itol annotation, Edson's Tree

#Load libraries
library(readr)
library(dplyr)
library(stringr)
```

```r
#Upload data
#representative genomes
 all_representative <- read_delim("microbiota_sialylation/genomes_download/plots_data/itol/representative_species_modified_v6.txt",  delim = "\t", col_names = F)
colnames(all_representative) <- "target_ID"
 
kpsT_itol_represent <-read_delim("microbiota_sialylation/genomes_download/plots_data/itol/kpsT_represent_filter.tsv",
                       delim = "\t")


kpsM_itol_represent <-read_delim("microbiota_sialylation/genomes_download/plots_data/itol/kpsM_represent_filter.tsv",
                                 delim = "\t")
oacetil_itol_represent <-read_delim("microbiota_sialylation/genomes_download/plots_data/itol/oacetil_represent_filter.tsv",
                                 delim = "\t")

vfdb_itol_represent <- read_delim("microbiota_sialylation/genomes_download/plots_data/itol/out_70id_60cov.csv",
                                  delim = ",")

rfah_itol_represent <- read_delim("microbiota_sialylation/genomes_download/plots_data/itol/rfah_represent_filter.tsv",
                                  delim = "\t")
```

```r
#for kpsT

#Extract ID and put in another columm
df_all <- all_representative %>%
  mutate(Part = str_extract(target_ID, "^[^_]*_[^_]*"))

#select first columm
kpsT_itol_represent <- kpsT_itol_represent[,1]

#Extract ID and put in another columm
df_kpsT <- kpsT_itol_represent %>%
  mutate(Part = str_extract(target_ID, "^[^_]*_[^_]*"))

#See what is different: the ones which were not in protein dataset's results
diff_representative_kpsT <- setdiff(df_all %>% select(Part), df_kpsT %>% select(Part))
head(diff_representative_kpsT)

#Atributte color and attach string "_protein" to IDs which were not identified any protein
diff_representative_kpsT$level <- "#F88379"
diff_representative_kpsT$Part <- paste0(diff_representative_kpsT$Part, "_protein")

head(diff_representative_kpsT)

#Atributte color and attach string "_protein" to ID's identified with protein
df_kpsT$level <- "#9F2B68"
df_kpsT$Part <- paste0(df_kpsT$Part, "_protein")
df_kpsT <- df_kpsT[ , -1]

head(df_kpsT)
```

```r
#Join the data
data_kpsT_itol_repre <- rbind(df_kpsT, diff_representative_kpsT)
head(data_kpsT_itol_repre)

#save result into a file
write_tsv(data_kpsT_itol_repre, "kpsT_represent_itol_EANS.tsv")
```

```r
#vfdb

#select first columm
vfdb_itol_represent <- vfdb_itol_represent[,1]
colnames(vfdb_itol_represent) <- "target_ID"

#take fisrt hit for each ID
vfdb_itol_represent <- distinct(vfdb_itol_represent)

#Extract ID and some part put in another columm
df_vfdb <- vfdb_itol_represent %>%
  mutate(Part1 = str_extract(target_ID, "^[^_]*_[^_]*"))

#Extract only the ID and put in another columm
df_vfdb1 <- df_vfdb %>%
  mutate(Part = str_extract(Part1, "(?<=/)[^/]+(?=/)"))

#see what is different between datasets
diff_representative_vfdb <- setdiff(df_all %>% select(Part), df_vfdb1 %>% select(Part))
head(diff_representative_vfdb)

#Atributte color and attach string "_protein" to IDs which were not identified any protein
diff_representative_vfdb$level <- "#DC143C"
diff_representative_vfdb$Part <- paste0(diff_representative_vfdb$Part, "_protein")
head(diff_representative_vfdb)

#Atributte color and attach string "_protein" to ID's identified with protein
df_vfdb1$level <- "#50C878"
df_vfdb1$Part <- paste0(df_vfdb1$Part, "_protein")
df_vfdb1 <- df_vfdb1[ , c(-1,-2)]

head(df_vfdb1)
```

```r
#label for phylum

host_data_comm_sia <- read_delim("host_data_comm_sia.tsv", 
                                 delim = "\t", escape_double = FALSE, 
                                 trim_ws = TRUE)
head(host_data_comm_sia)

#get_taxa_ID <- host_data_comm_sia[,1]

colnames(host_data_comm_sia)[colnames(host_data_comm_sia) == "Assembly Accession" ] <- "target_ID"
see_ID_comm <- inner_join(host_data_comm_sia, all_representative, by = "target_ID")


select_phylum <- see_ID_comm%>% select("target_ID", "Phylum name")

unique(select_phylum$`Phylum name`)
select_phylum$X1 <- paste0(select_phylum$target_ID, "_protein")

select_phylum$label_type <- "label_background"

#Atribbute color to matched phylum
cor_names <- c("Pseudomonadota" = "#8bd2c6", "Actinomycetota" = "#ffffb1", "Campylobacterota" = "#bdb9d9",
               "Bacteroidota" = "#fb7f71", "Bacillota"= "#fccce4", "Cyanobacteriota" = "#A5FFD6",
               "Fusobacteriota"= "#bb7fbc", "Thermodesulfobacteriota"= "#cbeac4", 
               "Deinococcota"= "#7fb0d2", "Spirochaetota" = "#b2dd69")

#for matched phylum, put hexcode color into another columm "COR_PHYLUM
select_phylum$COR_PHYLUM <- ifelse(as.character(select_phylum$`Phylum name`) %in% names(cor_names),
                                        cor_names[as.character(select_phylum$`Phylum name`)], "NULL")
#remove undesired columm
select_phylum <- select_phylum[,-2]

#save result into a file
write_tsv(select_phylum, "representative_phylum_itol.tsv")
```
