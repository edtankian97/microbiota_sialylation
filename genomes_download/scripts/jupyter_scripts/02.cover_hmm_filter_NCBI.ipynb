```r
#upload data for CMP-NeuA synthetase. Change PATH of the file, if needed
View_cmp_NCBI <-read.table("plots_data/CMP_coverage.tsv", 
                           header = FALSE, col.names = c("V1", "V2"))

head(View_cmp_NCBI)
nrow(View_cmp_NCBI)
```

```r
#cover filter

CMP_cover_filter_2_NCBI <- View_cmp_NCBI %>% filter(View_cmp_NCBI$V1 >= 40.00)

head(CMP_cover_filter_2_NCBI)
nrow(CMP_cover_filter_2_NCBI)
```

```r
# Save result into a file. Results will be inside HMMER_analysis folder

write_tsv(CMP_cover_filter_2_NCBI, "HMMER_analysis/CMP_NCBI_ID_cover_filter_40.tsv")
```
